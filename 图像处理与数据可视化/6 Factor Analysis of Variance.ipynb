# 6 Factor Analysis of Variance

By "factor" we mean a **discrete** covariate (e.g. categories) and by "factor level" a particular value (category) of factor. 




## Single Factor ANOVA

ANOVA is abbreviation for "analysis of variance".


### Model

Assume we are focusing on a factor $A$, which has $r$ different categories, indexed by $A_1,A_2,\dotsc,A_r$. For factor being the $A_i$, we have observations $y_{i1},y_{i2},\dotsc,y_{i,n_i}\in\mathbb R $. Assume these observations are from $N(\mu_i,\sigma^2)$.

In total there are $n_1+\dotsc+n_r=n$ observations. The variance $\sigma ^2$ is independent with the factor and the noises are uncorrelated.

### Notations

Sample mean of class $A_i$ is 
$$\bar y_{i\cdot} = \frac {1}{n_i}\sum_{j=1}^{n_i}y_{ij}=\frac{1}{n_i}\sum_{j=1}^n(\mu_i + \epsilon_{ij}) = \mu_i +\bar \epsilon_{i\cdot}.
$$

The sample mean of all data is
$$\bar y_{\cdot \cdot} = \frac {1}{n}\sum_{i=1}^r\sum_{j=1}^{n_i}y_{ij}=\left(\frac {1}{n}\sum_{i=1}^rn_i\mu_i \right)+\bar{\epsilon_{\cdot \cdot}}.
$$

### Estimators 

For each $i$, it is clear that the least squares estimator or MLE for mean is given by $\hat\mu_i = \bar y_{i\cdot}$. For $\sigma^2$, the MLE should be:

$$\begin{aligned}\hat\sigma_{\rm MLE}^2 &= {\rm argmax} \prod_{i,j}\frac{1}{\sqrt{2\pi\sigma^2}}\exp\left\{-\frac {1}{2\sigma^2} (y_{ij}-  \bar y_{i\cdot} )^2\right\}={\rm argmax}\left\{-\frac {1}{2\sigma^2} \sum _{ij}(y_{ij}-    \bar y_{i\cdot} )^2-\frac12n\log \sigma^2\right\}\\ &= \frac 1n \sum_{ij}(y_{ij} - \bar y_{i\cdot})^2.\end{aligned}
$$

However, the MLE for $\sigma^2$ is biased.


### Sum of Squares 

Define the total sum of squares to be ${\rm SST} = \sum_{i,j}(y_{ij} - y_{\cdot \cdot})^2$, and then
$${\rm SST} = \sum_{i=1}^r \sum_{j=1}^{n_i}\left[(y_{ij} - \bar y_{i\cdot})^2+(\bar y_{i\cdot} - \bar y_{\cdot \cdot})^2+2(y_{ij} - \bar y_{i\cdot})(\bar y_{i\cdot} - \bar y_{\cdot \cdot})\right]=\sum_{i=1}^r \sum_{j=1}^{n_i}\left[(y_{ij} - \bar y_{i\cdot})^2+(\bar y_{i\cdot} - \bar y_{\cdot \cdot})^2\right].$$

We then define ${\rm SSE} = S_e = \sum_{i=1}^r \sum_{j=1}^{n_i}(y_{ij} - \bar y_{i\cdot})^2$ and ${\rm SSR} = S_A = \sum_{i=1}^r \sum_{j=1}^{n_i}(\bar y_{i\cdot} - \bar y_{\cdot \cdot})^2=n_i\sum_{i=1}^r(\bar y_{i\cdot} - \bar y_{\cdot \cdot})^2$. We can see that ${\rm SSE}$ stands for the within-group error, while ${\rm SSR}$ is the between-group variation.



#### Distribution

**Theorem**   ${\rm S_e}\sim \sigma^2\chi_{n-r}^2$. And $S_A$ and $S_e$ are independent.

**Proof** 
For $S_e$, as we know that for all fixed $i$ we can find $\sigma^2\chi_{n_i-1}^2 =\sum_{j=1}^{n_i}(y_{ij} - \bar y_{i\cdot})^2 $ such that $\chi_{n_i-1}^2$ is independent with $\bar y_{i\cdot}$, so
$$ S_e \sim \sum_{i=1}^r  \sigma^2 \chi_{n_i-1}^2=\sigma^2 \chi_{n-r}^2  \perp\!\!\!\!\!\!\perp \sum_{i=1}^r n_i(\bar y_{i\cdot} - \bar y_{\cdot \cdot})^2=S_A.$$

As a consequence, $\hat \sigma^2 = \frac{\rm SSE}{n-r}$ gives an unbiased estimation for variance.


### Hypothesis Testing

#### Null Model

Assume we want to test $H_0:\ \mu_1=\mu_2=\dotsc=\mu_n$, meaning that factor A does not contribute. We can reject $H_0$ when $S_e$ is small enough, or equivalently $S_A$ is large enough. 

**Theorem** When $H_0$, we have $\frac{ S_A/(r-1)}{S_e / (n -r)} \sim \frac{\sigma^2\chi_{r-1}^2/(r-1)}{\sigma^2\chi_{n-r}^2/(n-r)} =F_{r-1,n-r}$ as the test statistics. 


**Proof** We only need to show that $S_A\sim \sigma^2\chi_{r-1}^2$ in this case. First,

$$S_A =\sum_{i=1}^r n_i(\bar \epsilon_{i\cdot} - \bar \epsilon_{\cdot \cdot})^2=\sum_{i=1}^r (\sqrt{n_i}\bar \epsilon_{i\cdot} - \sqrt{n_i}\bar \epsilon_{\cdot \cdot})^2.$$

Change the variable $z_i = \sqrt{n_i}\bar\epsilon_{i\cdot}\sim N(0,\sigma^2)$ and let $w = [\sqrt{n_i}/\sqrt n]_i^T \in\mathbb R^r$, then $w^Tw = 1$ and 

$$\bar \epsilon_{\cdot \cdot}
=\frac 1n \sum_{i=1}^r n_i \bar \epsilon_{i\cdot}=\frac 1n \sum_{i=1}^r \sqrt{n_i} z_i =\frac{1}{ \sqrt n} w^Tz.
$$

Let $\mathbf{1}_r = [1,1,\dotsc,1]^T\in\mathbb R^r$, then 
$$S_A = \Vert z - ww^Tz\Vert^2  = \Vert (I - ww^T)z\Vert^2.$$ 

Note that $(I - ww^T)$ is idemopotent with rank $(r-1)$, we learn $S_A\sim \sigma^2\chi_{r-1}^2$.

### Linear Regression

The model above can be related to linear regression. If we denote $e_i$ to be the $i$-th column of the identity matrix $\mathbb I_r$ while $\mu = [\mu_1,\dotsc,\mu_r]^T$, then the model is 
$$y_{ij} = e_i^T\mu + \epsilon_{ij}.$$

### Factor Level Mean 

If we require a confidence interval for $\mu_i$, we can note that $\bar y_{i\cdot } \sim N(\mu_i, \sigma^2/n_i)$, and $\bar y_{i\cdot} \perp\!\!\!\perp \bar {\rm SSE}$. So 
$$\frac{\sqrt n(\bar y_{i\cdot} - \mu_i)}{\sqrt{\frac{\rm SSE}{n-r}}}\sim t_{n-r}.$$

<br>

Similarly, if we want to derive a confidence interval for between-class difference $\mu_i - \mu_j$, we note that $\bar y_{i\cdot } - \bar y_{j\cdot}\sim N(\mu_i - \mu_j, \sigma^2(1/n_i+1/n_j))\perp \!\!\!\perp {\rm SSE}$ and thus 
$$\frac{\sqrt{\frac{n_in_j}{n_i+n_j}}\left[(y_{i\cdot} - y_{j\cdot}) - (\mu_i - \mu_j)\right]}{\sqrt{\frac{\rm SSE}{n-r}}}\sim t_{n-r}.$$

<br>

More generally, for $c\in\mathbb R^r$ and $c^T\mu = \sum c_i\mu_i$, we have $c^T\bar y_{i\cdot}\sim N\left(c^T \mu, \sigma^2 \sum \frac{c_i^2}{n_i}\right)$ and 
$$\frac{c^T\bar y_{i\cdot} - c^T  \mu}{\sqrt{\sum \frac{c_i^2}{n_i}}\cdot \sqrt{\frac{\rm SSE}{n-r}}}\sim t_{n-r}.$$

## Multiple Factors ANOVA

Assume now there are two factors, $A$ and $B$. $A$ has factor levels $A_1,\dotsc,A_a$ while $B$ has factor levels $B_1,\dotsc,B_b$. For joint factor levels $A_iB_j$, there are samples $y_{ij1},y_{ij2},\dotsc, y_{ij n}$ from $N(\mu_{ij},\sigma^2)$. And for all $(i,j)$, there are $n$ observations.

On true means, we denote $\bar \mu_{i\cdot  } = \frac 1b \sum_{j} \mu_{ij}$ and $\bar \mu_{\cdot j } = \frac 1a \sum_{ij}  \mu_{ij}$ and $\bar \mu_{\cdot \cdot} = \frac{1}{ab}\sum_{k}\mu_{ij}$.

On samples, we denote $\bar y_{i\cdot\cdot } = \frac {1}{bn}\sum_{jk} y_{ijk}$ and $\bar y_{\cdot j\cdot } = \frac {1}{an}\sum_{ijk}  y_{ijk}$ and $\bar y_{\cdot \cdot \cdot} = \frac{1}{abn}\sum_{k}y_{ijk}$.

For each single observation, we denote $\epsilon_{ijk} = y_{ijk} - \mu_{ji}$ by the noise. And we define in an analogue way that $\bar \epsilon_{i\cdot\cdot } = \frac {1}{bn}\sum_{jk} \epsilon_{ijk}$ and $\bar \epsilon_{\cdot j\cdot } = \frac {1}{an}\sum_{ijk}  \epsilon_{ijk}$ and $\bar \epsilon_{\cdot \cdot \cdot} = \frac{1}{abn}\sum_{k}\epsilon_{ijk}$


### Estimators

Similarly, $\hat\sigma_{\rm MLE}^2 = \frac{1}{abn}\sum (y_{ijk} - \bar y_{ij\cdot})^2$ and the unbiased estimator should be $\hat\sigma ^2 = \frac{1}{ab(n-1)}\sum (y_{ijk} - \bar y_{ij\cdot})^2$.

### Main Effect

Define the main effect of factor level $A_i$ by $a_i = \bar \mu_{i\cdot} - \bar \mu_{\cdot \cdot}$. Similarly, define the main effect of factor level $B_j$ by $b_j = \mu_{\cdot j} - \bar \mu_{\cdot \cdot}$. Define the treatment main effect $r_{ij}$ by 
$$r_{ij}  =\mu_{ij} - \bar \mu_{\cdot \cdot} - a_i - b_j = \mu_{ij} - \bar \mu_{i\cdot} - \bar \mu_{\cdot j} +\bar \mu_{\cdot \cdot}.$$

When all $r_{ij}=0$, we say that the effects $A,B$ are additive. If not, the model is said to have interacting effects.

<br>

To estimate $a_i,b_j,r_{ij}$, we can directly use the unbiased estimators:
$$\hat a_i = \bar y_{i\cdot\cdot}-\bar y_{\cdot \cdot\cdot}\quad \hat b_j = \bar y_{\cdot j\cdot}-\bar y_{\cdot \cdot\cdot} \quad \hat r_{ij} =\bar y_{ij\cdot}-\bar y_{i\cdot\cdot}-\bar y_{\cdot j\cdot}+\bar y_{\cdot \cdot\cdot} .$$

```python
from IPython.display import display

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
print('An example of additive effect:')
data = np.array([[11,13,18],[7,9,14]])
data = np.hstack([data, data.mean(axis = 1, keepdims = True)])
data = np.vstack([data, data.mean(axis = 0, keepdims = True)])
display(pd.DataFrame(data, columns = ['B1', 'B2', 'B3', 'Mean'], index = ['A1', 'A2', 'Mean']))

plt.plot(data[:-1, :-1].T, marker = '+', linewidth = 1)
plt.legend(['A%i'%i for i in range(1, data.shape[0])])
plt.xlabel('Factor B')
plt.title('Additive Effects Are Parallel Lines')
plt.show()
```

### Sum of Squares

Total sum of squares is independent with our model: ${\rm SST} = \sum_{ijk}(y_{ijk} - \bar y_{\cdot \cdot \cdot})^2$. While $S_e = \sum_{ijk}(y_{ijk} -\bar  y_{ij\cdot})^2$ is the error and $S_{AB} = \sum_{ijk}(\bar y_{ij\cdot} - \bar y_{\cdot \cdot})^2$.

Still we have ${\rm SST} = S_e + S_{AB}$ because $\sum_{ijk}(y_{ijk} -\bar  y_{ij\cdot})(\bar y_{ij\cdot} - \bar y_{\cdot \cdot})=0$.



#### Sab

Further, we can decompose $S_{AB}$:

**Theorem** $S_{AB} = \sum_{ijk}\left( \hat a_i^2+ \hat b_j^2 + \hat r_{ij}^2\right)={\rm SSA}+{\rm SSB}+{\rm SSAB}$.

**Proof** Note that $\hat a_i +\hat b_j +\hat r_{ij} = \bar y_{ij\cdot} - \bar y_{\cdot \cdot \cdot}$. So

$$S_{AB} = \sum_{ijk}\left(\hat a_i+\hat b_j+\hat r_{ij}\right)^2$$

Also, $\sum \hat a_i\hat b_j = \sum \hat b_j\hat r_{ij} = \sum \hat r_{ij}\hat a_i = 0$, so we conclude the result.


#### Degree of Freedom

We can see that ${\rm df}({\rm SST} )= abn - 1$,  ${\rm df}(S_e) = ab(n-1)$ and ${\rm df}(S_{AB}) = ab-1$.

Also, ${\rm df}({\rm SSA}) = a-1$, ${\rm df}({\rm SSB}) = b-1$ and finally
$${\rm df}({\rm SSAB}) = (ab-1)-(a-1)-(b-1) = (a-1)(b-1).$$

**Theorem** $S_e \sim \sigma^2\chi_{ab(n-1)}^2$. So $\hat\sigma^2 = \frac{S_e}{ab(n-1)}$ unbiasedly estimates the variance.

**Proof** Note that for all $(i,j)$, $\sum_{k=1}^n (\bar y_{ij\cdot} - y_{ijk})^2 \sim \sigma^2\chi_{n-1}^2$. And these $\chi_{n-1}^2$ are independent across different $(i,j)$.

$$S_e = \sum_{ijk}(\bar y_{ij\cdot } - y_{ijk})^2\sim \sum_{ij}\sigma^2\chi_{n-1}^2 = \sigma^2\chi_{ab(n-1)}^2.$$


<br>

**Theorem** $S_e$ is independent with ${\rm SSA}, {\rm SSB}, {\rm SSAB}$ and $S_{AB}$.

**Proof** Recall that we can for each $(i,j)$ find $\sigma^2\chi_{n-1}^2 = \sum_{k}(y_{ijk} - \bar y_{ij\cdot})^2$ such that $\chi_{n-1}^2$ is independent with $\bar y_{ij\cdot}$ , so we conclude that $S_e$ is independent with each $\bar y_{ij\cdot}$ and therefore independent with ${\rm SSA}, {\rm SSB}, {\rm SSAB}$ and $S_{AB}$.

### Hypothesis Testing

#### Test for Interaction

**Theorem** When assuming $H_0$: $r_{ij}=0$ for all $(i,j)$, (which is to say the factors are additive), then ${\rm SSAB}\sim \sigma^2\chi_{(a-1)(b-1)}^2$.

**Proof** When assuming $H_0$, the $\hat r_{ij}$ gets degenerated to the combination of noise:

$$\begin{aligned}\hat r_{ij} &= \bar y_{ij\cdot } -\bar y_{i\cdot \cdot } -\bar y_{\cdot j\cdot}+\bar y_{\cdot \cdot \cdot}
=(\bar \mu_{ij}+\bar \epsilon_{ij\cdot }) -(\bar \mu_{i\cdot}+\bar \epsilon_{i\cdot \cdot }) -(\bar \mu_{\cdot j}+\bar \epsilon_{\cdot j\cdot})+(\bar \mu_{\cdot\cdot}+\bar \epsilon_{\cdot \cdot \cdot})
\\ &= \bar \epsilon_{ij\cdot} - \bar\epsilon_{i\cdot \cdot} - \bar\epsilon_{\cdot j \cdot}+\bar \epsilon_{\cdot\cdot\cdot}
\end{aligned}
$$

Thus,
$${\rm SSAB} = \sum_{ijk}\hat r_{ij}^2 = n\sum_{ij}\hat r_{ij}^2
= n\sum_{ij}(\bar \epsilon_{ij\cdot} - \bar\epsilon_{i\cdot \cdot} - \bar\epsilon_{\cdot j \cdot}+\bar \epsilon_{\cdot\cdot\cdot})^2.$$

Now we let $\mathcal E  = [\bar \epsilon_{ij\cdot }]\in\mathbb R^{a\times b}$ to be the matrix. If we write $V =[\bar \epsilon_{ij\cdot} - \bar\epsilon_{i\cdot \cdot} - \bar\epsilon_{\cdot j \cdot}+\bar \epsilon_{\cdot\cdot\cdot}]_{ij}$ and denote $\mathbf{1}_r = [1,\dotsc,1]^T\in\mathbb R^r$, then we can see that 
$$\begin{aligned}V &=\mathcal E - \mathcal E \frac 1b \mathbf{1}_b\mathbf{1}_b^T-\frac 1a  \mathbf{1}_a\mathbf{1}_a^T\mathcal E +\frac{1}{ab}\mathbf{1}_a\mathbf{1}_a^T\mathcal E \mathbf{1}_b\mathbf{1}_b^T
=(I - \frac 1a\mathbf{1}_a\mathbf{1}_a^T)\mathcal E(I - \frac 1b\mathbf{1}_b\mathbf{1}_b^T)
\\ \Rightarrow {\rm vec}(V) &= \left[(I - \frac 1b\mathbf{1}_b\mathbf{1}_b^T)\otimes (I - \frac 1a\mathbf{1}_a \mathbf{1}_a^T)\right]{\rm vec}(\mathcal E)
\end{aligned}$$


<!-- <br> -->

Note that $\bar \epsilon_{ij\cdot }$ are independent noise from $N(0,\sigma^2/n)$ across $(i,j)$ and the Kronecker product $\left[(I - \frac 1b\mathbf{1}_b\mathbf{1}_b^T)\otimes (I - \frac 1a\mathbf{1}_a \mathbf{1}_a^T)\right]$ is also idempotent with rank $(a-1)(b-1)$, so we know that 
$${\rm SSAB} = n\Vert V\Vert_F^2 \sim \sigma^2\chi_{(a-1)(b-1)}^2. $$


<br>

To conclude, we can test 
$$\frac{{\rm SSAB}/((a-1)(b-1))}{S_e / (ab(n-1))}\sim F_{(a-1)(b-1),ab(n-1)}.$$

We reject $H_0$ when the left hand side is abnormally large.

#### Testing Main Effect Without Interaction

If we want to test $H_0:\ a_1=a_2=\dotsc=a_a = 0$, i.e. the main effects of factor $A$ are strictly zero.

Assuming there is no interaction between factors $A,B$ (since we have assumed $A$ has no effect, there cannot be any interaction): $r_{ij}=0$ for all $(i,j)$, then we have shown that in this case ${\rm SSAB}$ does not depend on any $\mu_{ij}$ but merely the noise:
$${\rm SSAB} = n\sum_{ij}(\bar \epsilon_{ij\cdot} - \bar \epsilon_{i\cdot \cdot} - \bar \epsilon_{\cdot j\cdot}+\bar \epsilon_{\cdot \cdot\cdot})^2$$

Therefore, we should  include ${\rm SSAB}$ when estimating the variance (so that it would be more accurate):
$$\tilde S_e = S_e + {\rm SSAB}\sim \sigma^2\chi_{abn-a-b+1}^2\quad\quad \hat\sigma^2 =\frac{\tilde S_e}{abn-a-b+1}.$$

<br>

**Theorem** Assuming there is no interaction and $H_0$, then the test statistics is given by $\frac{{\rm SSA}/(a-1)}{{\tilde{S_e}}/(abn-a-b+1)}\sim F_{a-1,abn-a-b+1}$.

**Proof** When $H_0$, it is clear that ${\rm SSA}\sim \sigma^2\chi_{a-1}^2$. It suffices to prove that ${\rm SSA}$ and $\tilde{S_e}$ are independent in this case. As ${\rm SSA}$ and $S_e$ are independent, we only need to show that ${\rm SSA}$ and ${\rm SSAB}$ are independent.

Still we use the notation $\mathcal E = [\bar  \epsilon_{ij\cdot}]_{ij}\in\mathbb R^{a\times b}$, and recall that 
$${\rm SSAB} = \left\Vert \left[(I - \frac 1b \mathbf{1}_b\mathbf{1}_b^T)\otimes (I - \frac 1a \mathbf{1}_a \mathbf{1}_a^T)\right]{\rm vec}(\mathcal E)\right\Vert^2.$$

Here $\mathbf{1}_r = [1,1,\dotsc,1]^T\in\mathbb R^r$. Similarly we can write

$$\begin{aligned}{\rm SSA} &= \sum_{ijk}(\bar y_{i\cdot\cdot} - \bar y_{\cdot \cdot \cdot})^2
=n\sum_{ij} (\bar y_{i\cdot \cdot} - \bar y_{\cdot \cdot\cdot})^2
=n \sum_{ij} (\bar \epsilon_{i\cdot\cdot} - \bar \epsilon_{\cdot\cdot\cdot})^2
\\ &= \left\Vert \mathcal E (I - \frac 1b \mathbf{1}_b\mathbf{1}_b^T) - (I - \frac 1a \mathbf{1}_a \mathbf{1}_a^T) \mathcal E (I - \frac 1b \mathbf{1}_b\mathbf{1}_b^T) \right\Vert_F^2
\\ &=  \left\Vert \left[(I - \frac 1b \mathbf{1}_b\mathbf{1}_b^T)\otimes  \frac 1a \mathbf{1}_a \mathbf{1}_a^T\right]{\rm vec}(\mathcal E)\right\Vert^2.
\end{aligned}
$$

Finally, note that $\left[(I - \frac 1b \mathbf{1}_b\mathbf{1}_b^T)\otimes  \frac 1a \mathbf{1}_a \mathbf{1}_a^T\right]$ is idempotent and orthogonal to $ \left[(I - \frac 1b \mathbf{1}_b\mathbf{1}_b^T)\otimes (I - \frac 1a \mathbf{1}_a \mathbf{1}_a^T)\right]$, the independence follows from Cochran's theorem.
