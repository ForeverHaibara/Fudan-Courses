# 3 拟合

由于数据存在误差, 龙格现象等原因, 不一定要求插值函数经过所有数值点, 则一般需要使得误差最小. 例如给定高维数据以及对应的函数值: $(x_1,y_1)\dotsc,(x_n,y_n)\in \mathbb R^m\times \mathbb R$. 称为拟合 (fitting).

## 最小二乘法

设所有 $n$ 个 $x_i\in\mathbb R^m$ 构成矩阵 $X\in \mathbb R^{n\times m}$.

一组基函数 $\phi_1,\dotsc,\phi_r:\ \mathbb R^m\times \mathbb R$ 并定义 
$$\mathcal V\stackrel{\triangle}{=}{\rm span}\{\phi_1,\dotsc,\phi_r\}$$
是 $\phi_i$ 的线性组合. 寻找一组系数 $c_i$ 并取 $\phi = c_1\phi_1 + \dotsc + c_r\phi_r$.

代入 $x=x_i$, 需要使得误差最小 (r < n)
$$y + \Delta y = \left[\begin{matrix}\phi_1(x_1) & \phi_2(x_1) & \dotsc & \phi_r(x_1)\\ 
\phi_1(x_2) & \phi_2(x_2) & \dotsc & \phi_r(x_2)\\
\vdots & \vdots & \ddots & \vdots\\
\phi_1(x_n) & \phi_2(x_n) & \dotsc & \phi_r(x_n)\end{matrix}\right]
\left[\begin{matrix}c_1\\c_2\\ \vdots \\ c_r\end{matrix}\right]=\Phi(X)c
$$

### 普通最小二乘 (OLS)

普通最小二乘 (OLS, ordinary least squares) 使得 $\Delta y$ 最小.
即 $$\min_c \Vert\Delta y\Vert = \min_c \Vert \Phi(X)c - y\Vert$$
容易解得理论解(之一) $c = \Phi(X)^\dag y $.

### 总体最小二乘 (TLS)

总体最小二乘 (TLS, total least squares) 考虑到 $x$ 也有扰动, 即 $y + \Delta y = \Phi(X+\Delta X)c$, 且需要使得
$\Vert [\Delta x, \Delta y]\Vert$ 最小. 

为了方便, 只考虑 $\phi_1,\dotsc,\phi_r$ 是线性无关的线性函数, $\phi_i (x) = x^Tv_i:\ \mathbb R^m\rightarrow \mathbb R$. 则
$$y+\Delta y = (X+\Delta X)Vc$$
抑或写成
$$(\left[\begin{matrix}X & y\end{matrix}\right]+\left[\begin{matrix}\Delta X &\Delta y\end{matrix}\right] ) 
\left[\begin{matrix}Vc \\ -1\end{matrix}\right]= 0$$
这表明矩阵 $\left[\begin{matrix}X & y\end{matrix}\right]$ 经过微小扰动 (perturbation) $\left[\begin{matrix}\Delta X & \Delta y\end{matrix}\right]$ 后不满秩, 根据最佳低秩逼近
$$\min \Vert \left[\begin{matrix}\Delta X & \Delta y\end{matrix}\right]\Vert_2
=\sigma_{r+1}\left(\left[\begin{matrix}X & y\end{matrix}\right]\right)$$



## 非线性拟合

在上述最小二乘法 (线性拟合) 过程中, 可以视为构造了多元函数 $\Phi(x,c) = \sum_{k=1}^r c_k\phi_k(x)$ 并求解
$$\min_c \sum_{i=1}^n \Vert y_i - \Phi(x_i,c)\Vert^2 $$
由于 $\frac{\partial ^2}{\partial c^2}\Phi(x,c) \equiv 0$, 故 $\Phi(x,c)$ 是关于 $c$ 的线性问题, 称为线性插值 / 拟合 . 更一般地, $\Phi(x,c)$ 可能不是关于 $c$ 的线性函数, 即非线性拟合.

### 高斯牛顿法

高斯-牛顿法 (Gauss-Newton / GN method) 可以做非线性拟合问题. 一般地, 对于模型 (model) 
$$\hat y = \Phi (X,c):\ \mathbb R^{n\times m}\times \mathbb R^p \rightarrow \mathbb R^n$$
是关于 $n$ 个输入 (inputs) $X\in \mathbb R^{n\times m}$ 与参数 (parameters) $c$ 的非线性函数. 其中 $X$ 的每行是一个 $\mathbb R^m$ 的数据向量. $y\in \mathbb R^n$ 则是对应的 $n$ 个输出. 数值求解此系统可以作如下迭代:
$$c^{(k+1)}=c^{(k)}+\left(J^{(k)}\right)^\dag r^{(k)}$$

其中 $r^{(k)} = y - \Phi (X,c)$ 为残量. $J^{(k)} = \frac{\partial \Phi}{\partial c^{(k)}}\in \mathbb R^{p\times n}$ 为梯度. 符号
$\left(J^{(k)}\right)^\dag \in \mathbb R^{n\times p}$ 表示其伪逆 (pseudo-inverse). 伪逆乘向量可以等价于求解最小二乘法问题
$$\left(J^{(k)}\right)^\dag r^{(k)} = \argmin_{x}\Vert J^{(k)}x - r^{(k)}\Vert_2^2$$

高斯-牛顿法是类似牛顿法的迭代法, 具有局部二次收敛性, 但不一定全局收敛.


#### 阻尼高斯-牛顿法

带阻尼的高斯-牛顿法 (Damped Gauss-Newton) 只是在每次 $c^{(k)}$ 更新中乘上一个系数

$$c^{(k+1)}=c^{(k)}+\alpha^{(k)}\left(J^{(k)}\right)^\dag r^{(k)}$$

可以选取 $\alpha$ 满足 Wolfe 条件.

#### Levenberg-Marquardt

若梯度 $J$ 是列满秩的, 则熟知伪逆有公式 $J^\dag = (J^TJ)^{-1}J^T$. 

Levenberg-Marquardt (LM) 方法则引入一个参数 $\lambda$ (称为 Tikhonov 参数) 与矩阵 $D$ 每次更新改为:

$$\Delta c = (J^TJ + \lambda D)^{-1}J^Tr$$

#### 拟牛顿法

拟牛顿法 (Quasi-Newton) 提出类似切线法的近似梯度矩阵 $J$ 的方法避免求导.

## 最小二乘逼近

更一般地, 最小二乘逼近 (least squares approximation) 需要函数在区间上逼近而非只在数据点上拟合, 即最小化某个函数二范数
$$\min_{\phi \in \mathcal V}\Vert f(x) - \phi(x)\Vert_2^2$$

其中 $\Vert \cdot \Vert_2$ 是定义在函数内积空间 $\mathcal U$ 的二范数. 二范数是内积诱导的. $\mathcal V$ 是 $\mathcal U$ 的子空间, 表示目标 $\phi (x)$ 只在该子空间内选取. 

特别地, 若在该内积定义下 $\mathcal V$ 可以选取一组标准正交基 (orthonormal basis) $\phi_1(x),\dotsc,\phi_n(x)$, 则最优的 $\phi_*(x)$ 由如下式子给出

$$\phi_*(x) = \sum_{i=1}^n \langle f(x),\phi_i(x)\rangle  \phi_i(x)$$

证: 设 $\phi(x) = \sum_{i=1}^n a_i \phi_i(x)$, 则
$$\begin{aligned}
\Vert f(x) - \phi(x)\Vert^2&=\Vert f(x)\Vert^2 - \langle f(x),\phi(x)\rangle - \langle \phi(x),f(x)\rangle + \Vert \phi(x)\Vert^2\\
&= \Vert f(x)\Vert^2 - 2\Re \left(a_i \langle \phi_i(x),f(x)\rangle\right) + \sum_{i=1}^n |a_i|^2\\
&\geqslant \Vert f(x)\Vert^2+ \sum_{i=1}^n \left(\left|\overline{a_i} - \langle \phi_i(x),f(x)\rangle\right|^2 - \left|\langle \phi_i(x),f(x)\rangle\right|^2\right)\\
& \geqslant \Vert f(x)\Vert^2 - \sum_{i=1}^n  \left|\langle \phi_i(x),f(x)\rangle\right|^2\\
& = \Vert f(x)\Vert^2 - \Vert \phi_*(x)\Vert^2 
\end{aligned}
$$

最小值显然在 $a_i =  \langle f(x),\phi_i(x)\rangle$ 取到.


#### Bessel 不等式

上述过程表明, 
$$\sum_{i=1}^n |\langle f(x), \phi_i(x)\rangle|^2 = \Vert \phi_*(x)\Vert^2\leqslant \Vert f(x)\Vert^2$$

取等时 $f(x) - \phi_*(x)\equiv 0$, 即 $f(x)\in \mathcal V$.


#### 函数内积

通常给定 Lebesgue 可积的, 非负的权重函数 (weight function) $\rho(x)$, 可以依此定义函数内积为
$$\langle f(x),g(x)\rangle = \int_{-\infty}^{+\infty}\rho(x) f(x)\overline {g(x)}dx$$
特别地若 $\langle f(x),g(x)\rangle = 0$ 则称在该内积空间中函数 $f(x),g(x)$ 正交. 

记该内积空间为 $\mathcal U = L_{\rho}^2(\mathbb R)$. 

也可以仅在一个区间 $[a,b]$ 内积分, 
$$\langle f(x),g(x)\rangle = \int_{a}^{b}\rho(x) f(x)\overline {g(x)}dx$$

记该内积空间为 $\mathcal U = L_{\rho}^2[a,b]$. 

## 正交多项式

回到最小二乘法. 若在某函数内积空间上 $\{p_k(x)\}_{k=0}^{\infty}$ 为一组多项式且满足
$$\begin{aligned}
{\rm deg}\ p_n(x) &= n \\ 
\langle p_m(x),p_n(x) \rangle & = 0\quad (m\neq n)
\end{aligned}$$
则称 $p_0(x),p_1(x),\dotsc$ 是该内积空间上的一组正交多项式 (orthogonal polynomials). 特别地若内积采用 $$\langle f(x),g(x)\rangle = \int_{-\infty}^{+\infty}\rho(x) f(x)\overline {g(x)}dx$$
的定义, 可以称 $p_0(x),p_1(x),\dotsc$ 是权重 $\rho(x)$ 上的一组正交多项式. 即在 $L_{\rho}^2(\mathbb R)$ (或 $L_{\rho}^2[a,b]$) 的内积空间中考虑.

特别地, 若对所有 $k$ 还有 $\langle p_k(x),p_k(x)\rangle =1$, 则称它们是标准正交的 (orthonormal).

### 基

$\{p_k(x)\}_{k=0}^{\infty}$ 是多项式函数空间的一组基. 在接下来的推导中, 使用如下记号: 多项式 $p_n(x)$ 可以写成单项式的和,
$$p_n(x) = p_{0,n} + p_{1,n}x\dotsc+p_{n,n}x^n=[1,x,\dotsc,x^n][p_{0,n},p_{1,n},\dotsc,p_{n,n}]^T$$


对于不超过 $n-1$ 次的所有多项式构成的线性空间 $\mathcal V_n$, 记 $P_n = [p_{ij}]\in \mathbb C^{n\times n}$ 与 $X_n = [x_{ij}]\in \mathbb C^{n\times n}$, 在无混淆时分别简记为 $P$. 显然 $P$ 是上三角.

设该内积下 $V = [\langle x^i,x^j\rangle]_{0\leqslant i,j<n}$ 为 Gram 矩阵, 表示函数 $x^i$ 与 函数 $x^j$ 的内积. 那么存在对角阵 $D$, 
$$D=[\langle p_i(x),p_j(x)\rangle]_{0\leqslant i,j<n} = P^*VP.$$

显然 $V$ 应当具有如下实对称且类似 Toeplitz 的结构:

$$V = \left[\begin{matrix}v_0 & v_1 & v_2 &\dotsc & v_{n-1} \\
v_1 & v_2 & v_3 &\dotsc & v_n\\ 
v_2 & v_3 & v_4 & \dotsc & v_{n+1}\\
\vdots & \vdots & \vdots & \ddots & \vdots \\
v_{n-1} & v_n & v_{n+1} & \dotsc &v_{2n-2}
\end{matrix}\right]\in \mathbb R^{n\times n}$$

### 三项递推

若 $\{p_k(x)\}_{k=0}^{\infty}$ 为一组正交多项式, 则存在复数列 $\{\alpha_n\},\{\beta_n\},\{\gamma_n\}$, 
$$xp_n(x) = \gamma_np_{n-1}(x) + \alpha_n p_n(x) + \beta_n p_{n+1}(x)$$
这被称为三项递推性 (three-term recurrence). 它是一种 Arnoldi 过程 (详见数值算法与案例分析I 的 Krylov 子空间一节).

特别地, 若 $\{p_k(x)\}_{k=0}^{\infty}$ 为一组标准正交多项式, 则
$$xp_n(x) =\overline{ \beta_{n-1}}p_{n-1}(x) + \alpha_n p_n(x) + \beta_n p_{n+1}(x)$$
且 $\alpha_n\in\mathbb R$, 这是一种 (Hermite 对称的) Lanczos 过程.

证: 一方面, $[xp_{0}(x),\dotsc,xp_{n-1}(x)]$ 在基 $[1,x,\dotsc,x^n]$ 下的矩阵为 $E P_{n}$. 
其中 $E = \left[\begin{matrix}0 \\ I_{n}\end{matrix}\right]\in \mathbb C^{(n+1)\times n}$ 为乘 $x$ 的线性算子. 

另一方面, $[xp_{0}(x),\dotsc,xp_{n-1}(x)]$ 可以作为 $p_i(x)\ (i\leqslant n)$ 的线性组合 $P_{n+1}\widetilde H_n$. 

其中 $\widetilde H_n\in \mathbb C^{(n+1)\times n}$, 由于 $xp_i(x)$ 是 $(i+1)$ 次多项式, 是前 $(i+2)$ 个多项式的线性组合， 故显然 $\widetilde H_n$ 是上 Hessenberg 矩阵. 设 $\widetilde H_n = \left[\begin{matrix} H_n\\ 0 , \beta_n \end{matrix}\right]$, $P_{n+1} = [\widetilde P_n, p_{n+1}]\in\mathbb C^{(n+1)\times (n+1)}$, 
$\widetilde P_n = \left[\begin{matrix}P_n \\ 0\end{matrix}\right]\in \mathbb C^{(n+1)\times n}$,
$\widetilde E = \left[\begin{matrix}0 & 0\\ I_{n} & 0\end{matrix}\right]\in \mathbb C^{(n+1)\times (n+1)}$ 

$$\widetilde E \widetilde P_n = E P_{n}=P_{n+1}\widetilde H_n = \widetilde P_{n}H_n + p_{n+1}[0,\dotsc,0,\beta_n] $$

上式即 Arnoldi 过程. 注意正交性: $\widetilde P_n^* V_{n+1} \widetilde P_{n}  =D_n$, 以及 $\widetilde P_n^* V_{n+1} p_{n+1} = 0$.
$$ \widetilde P_n^* V_{n+1}  \widetilde E \widetilde P_{n}  =D_n\widetilde H_n$$
注意 $\widetilde P_n = \left[\begin{matrix}I_n \\ 0\end{matrix}\right]P_n$, 化简得
$$P_n^* \widehat{V_n} P_n= D_n\widetilde H_n$$
其中 $\widehat {V_n}$ 为 $V_{n+1}$ 右上角 $n\times n$ 的子矩阵. 由前面性质知 $\widehat {V_n}$ 对称. 从而上式暗示 
$\widetilde H_n$ 作为上 Hessenberg 矩阵是 Hermite 对称阵, 故其是 Hermite 对称的三对角. 特别地, 若其是标准正交的, $D_n$ 为单位阵, 递推也是 Hermite 三对角矩阵.


```python
import sympy as sp 
import numpy as np
n = 5
P = sp.Matrix.zeros(n+1,n+1)
P[0,0] = sp.sqrt(sp.Rational(1 , 2)) # normalize on [-1,1]
# Legendre polynomials
for i in range(1, n+1): # columns 
    tmp = sp.polys.polytools.Poly(sp.legendre_poly(i)).coeffs()
    for j, val in zip(range((i%2), (i%2)+2*len(tmp)+1, 2), tmp[::-1]): # rows
        P[j,i] = val
    # normalize
    P[:,i] *= sp.sqrt(sp.Rational((2 * i + 1) , 2))

# Matrix V: Gram matrix
V = sp.Matrix.zeros(n+1,n+1)
for i in range(n+1):
    for j in range(n+1):
        if i == j: V[i,j] = sp.Rational(2, i + i + 1)
        elif (i-j)%2 == 0: V[i,j] = sp.Rational(2, i + j + 1)
        else: V[i,j] = 0
P.T @ V @ P # orthonormal
```

```python
E = sp.Matrix.zeros(n+1,n+1)
E[1:,:-1] = sp.Matrix.eye(n,n)
P2 = P[:,:-1]
H = P2.T @ V @ E @ P2
H # the recurrence matrix should be tridiagonal
# and in particular, Hermitian for orthonormal polynomials
```

### 零点

若 $\{p_k(x)\}_{k=0}^{\infty}$ 为一组<font color = red>标准正交</font>多项式, 具有 Lanczos (Hermite) 递推公式:
$$xp_n(x) = \overline{\beta_{n-1}}p_{n-1}(x) + \alpha_np_n(x) +\beta_n p_{n+1}(x)$$

则 $p_n(x)$ 的零点是上述递推的 n 阶三对角 Hermite 矩阵的所有特征值. 且零点都是实数且互不相同.
$$\left[\begin{matrix}\alpha_0 &\beta_0 & & & & \\ 
 \overline{\beta_0} & \alpha_1 &  \beta_1 & & & \\
 &\overline{\beta_1} & \alpha_2 & \ddots & & \\  
 & & \ddots & \ddots &  \beta_{n-3}& \\ 
 & & &\overline{\beta_{n-3}} & \alpha_{n-2} & \beta_{n-2}\\
 & & & &  \overline{\beta_{n-2}} & \alpha_{n-1} 
\end{matrix}\right]$$

注: 根据标准正交性, 显然可用区间上的积分求出 $\alpha_n = \int_a^b \rho(x)xp_n^2(x)dx$ 以及 $\beta_n = \int_a^b \rho(x)xp_n(x)p_{n+1}(x)dx$.

证: 根据递推性, 注意对于任意 $x\in \mathbb C$,
$$\left[\begin{matrix}p_0(x) \\ p_1(x) \\ p_2(x) \\ \vdots \\p_{n-2}(x)\\p_{n-1}(x)\end{matrix}\right]x = 
\left[\begin{matrix}\alpha_0 &\beta_0 & & & & \\ 
 \overline{\beta_0} & \alpha_1 &  \beta_1 & & & \\
 &\overline{\beta_1} & \alpha_2 & \ddots & & \\  
 & & \ddots & \ddots &  \beta_{n-3}& \\ 
 & & &\overline{\beta_{n-3}} & \alpha_{n-2} & \beta_{n-2}\\
 & & & &  \overline{\beta_{n-2}} & \alpha_{n-1} 
\end{matrix}\right]
\left[\begin{matrix}p_0(x) \\ p_1(x) \\ p_2(x) \\ \vdots \\p_{n-2}(x)\\p_{n-1}(x)\end{matrix}\right]
+
\left[\begin{matrix}0 \\ 0 \\ 0\\ \vdots \\0\\ \beta_np_n(x)\end{matrix}\right]
$$

可见只要 $p_n(x)=0 \Rightarrow x$ 就是特征值. 反之, 若 $x$ 是特征值, 设 $[a_0,a_1,\dotsc,a_{n-1}]$ 是对应的特征向量, 不妨设 $a_0 = p_0(x)$ (注意 $p_0(x)\neq 0$ 为常数次正交多项式, 且 $a_0 \neq 0$ 否则递推出零特征向量). 容易递推出特征向量恰好是 
$[p_0(x),p_1(x),\dotsc,p_{n-1}(x)]$. 由最后一个等式限制推知 $p_n(x) = 0$.


<br>


最后证明该三对角矩阵的 $n$ 个零点互不相同. 这其实是不可约对称三对角的性质: 显然根据递推公式, $\beta_n\neq 0$, 否则会在等式右边无 $n+1$ 次项. 所以对于 $\forall \lambda$, 矩阵
$$\left[\begin{matrix}\alpha_0-\lambda &\beta_0 & & & & \\ 
 \overline{\beta_0} & \alpha_1 -\lambda&  \beta_1 & & & \\
 &\overline{\beta_1} & \alpha_2 -\lambda& \ddots & & \\  
 & & \ddots & \ddots &  \beta_{n-3}& \\ 
 & & &\overline{\beta_{n-3}} & \alpha_{n-2}-\lambda & \beta_{n-2}\\
 & & & &  \overline{\beta_{n-2}} & \alpha_{n-1}-\lambda 
\end{matrix}\right]$$

显然左下角 $(n-1)\times (n-1)$ 的子式是上三角矩阵, 行列式非零, 所以整个矩阵的秩不低于 $n-1$, 那么零特征值几何重数不超过 $1$. 而这时 Hermite 矩阵, 故可以对角化所以代数重数与几何重数相等, 因此零特征值的代数重数不超过 $1$. 取 $\lambda$ 为所有特征值即得.

```python
deg = 5
# 将特征值代入多项式检查是否为 0
roots = np.linalg.eigvalsh(np.array(H).astype('float'))
f = sp.polys.polytools.Poly(sp.legendre_poly(deg))
for root in roots:
    print(float(f(root)))
```

### 交错性

若 $\{p_k(x)\}_{k=0}^{\infty}$ 为一组正交多项式, 设 $p_{n-1}(x)$ 所有零点是 $y_1<y_2<\dotsc<y_{n-1}$ 且 $p_{n}(x)$ 所有零点是 $x_1<x_2<\dotsc<x_n$, 则
$$x_1<y_1<x_2<y_2<\dotsc<x_{n-1}<y_{n-1}<x_n$$

证: 由于多项式乘若干倍不会改变零点, 不妨设它们是标准正交多项式. 根据 Hermite 的特征值性质, 
$$x_1\leqslant y_1\leqslant x_2\leqslant y_2\leqslant \dotsc\leqslant x_{n-1}\leqslant y_{n-1}\leqslant x_n$$
是柯西交错定理 (Cauchy's interlacing theorem). 下证明严格交错. 反证, 假设 $p_n(x)$ 与 $p_{n-1}(x)$ 有公共零点 $x$, 
则由于上一性质证明中已经指出过 $\beta_{n-1}\neq 0\ (n\geqslant 1)$, 我们有
$$p_{n-1}(x) = \frac{1}{\overline{\beta_{n-1}}}\left((x-\alpha_n)p_n(x) - \beta_n p_{n+1}(x)\right) = 0$$
递推下去, $p_{n-1}(x),\dotsc,p_0(x)$ 都有零点 $x$, 矛盾.


## 最佳一致逼近

最佳一致逼近 (best uniform approximation) 需要在区间内误差极大值最小化. 即
$$\inf_{\phi\in\mathcal V}\Vert f(x) - \phi(x)\Vert_{\infty}$$

特别地, 以下我们只考虑多项式的最佳一致逼近, 即目标函数 $f\in C[a,b]$, 给定函数空间
$\phi \in \mathcal V = \mathbb R_n[x] = {\rm span}\{1,x,x^2,\dotsc,x^n\}$, $\phi(x)$ 是不超过 $n$ 次的多项式. 求解
$$\min_{\phi \in\mathcal V} \Vert f(x) - \phi(x)\Vert_{\infty} = \min_{\phi \in\mathcal V} \sup_{x\in [a,b]}|f(x) - \phi(x)|$$

#### Weierstrass 最佳一致逼近定理

Weierstrass 最佳一致逼近定理称: 随着多项式次数升高, 紧集上可以做任意好的一致逼近:
$$\lim_{n\rightarrow\infty}\inf_{\phi \in \mathbb R_n[x]}\Vert f(x) - \phi(x)\Vert_{\infty}=0$$

#### 凸性

误差函数 $ \Vert f(x) - \phi(x)\Vert_{\infty}$ 是关于 $\phi$ 有下界的凸函数(convex function). 

证: 设 $\phi_1(x),\phi_2(x)\in\mathcal V$, $\alpha \in [0,1]$,
$$0\leqslant \Vert f(x) - \left(\alpha \phi_1(x) + (1 - \alpha)\phi_2(x)\right)\Vert_{\infty}
 \leqslant \alpha \Vert f(x) - \phi_1(x)\Vert_{\infty}
 +(1 - \alpha) \Vert f(x) - \phi_2(x)\Vert_{\infty}$$

#### 存在性

若 $\mathcal V$ 为有限维空间, 下式中下确界可取等, 即存在最小值.
$$\inf_{\phi\in\mathcal V}\Vert f(x) - \phi(x)\Vert_{\infty}$$

证: 若 $\Vert \phi(x)\Vert_{\infty}> 2\Vert f(x)\Vert_{\infty}$ 则
$$\Vert f(x) - \phi(x)\Vert_{\infty}\geqslant \Vert \phi(x)\Vert_{\infty}- \Vert f(x)\Vert_{\infty}
> \Vert f(x)\Vert_{\infty}  = \Vert f(x) - 0\Vert_{\infty}$$
说明此时选取 $\phi(x)$ 不如选取零多项式作最佳一致逼近. 故不妨只考虑 $\phi(x)$ 选取于紧集 $\Vert \phi(x)\Vert_{\infty}\leqslant \Vert f(x)\Vert_{\infty}$ 之中. 而 $\mathcal V$ 是有限维的完备赋范空间, 紧集上目标函数必有最小值.



### 交错点组

#### Chebyshev 交错点组

对于函数 $f(x)$ 在闭区间 $[a,b]$ 上, 若存在 $m$ 个点 $ a\leqslant x_1< x_2 < \dotsc x_m \leqslant b$ 使得
$$f(x_i) = (-1)^i \Vert f(x)\Vert_{\infty}\quad (i = 1,2,\dotsc,m )$$
或
$$f(x_i) = (-1)^{i-1} \Vert f(x)\Vert_{\infty}\quad (i = 1,2,\dotsc,m )$$
则称 $x_1,x_2,\dotsc,x_m$ 是一组 Chebyshev 交错点组 (Chebyshev alternating set).

#### 非均匀交错点组
对于函数 $f(x)$ 在闭区间 $[a,b]$ 上, 若存在 $m$ 个点 $ a\leqslant x_1< x_2 < \dotsc x_m \leqslant b$ 使得
$$(-1)^if(x_i) > 0\quad (i = 1,2,\dotsc,m )$$
或
$$(-1)^{i-1}f(x_i) > 0\quad (i = 1,2,\dotsc,m )$$
则称 $x_1,x_2,\dotsc,x_m$ 是一组 非均匀交错点组 (nonuniform alternating set).

#### de la Vallée-Poussin 定理

设 $f(x)\in C[a,b],\ \phi(x)\in \mathbb R_n[x]$, 若 $f(x) - \phi(x)$ 存在至少 $n+2$ 个点的非均匀交错点组, 则最佳一致逼近有下界
$$\min_{\Phi \in \mathbb R_n[x]}\Vert f(x) - \Phi(x)\Vert_{\infty}\geqslant \min_i \{|f(x_i) - \phi(x_i)|\}$$
证: 反证假设存在 $\phi_*(x)$ 使得 $\Vert f(x) - \phi_*(x)\Vert_{\infty}< \min_i \{|f(x_i) - \phi(x_i)|\}$ 于是
$$| f(x_i) - \phi_*(x_i)|< |f(x_i) - \phi(x_i)|$$
从而
$\phi_* (x_i) -  \phi (x_i)= \left(f(x_i) - \phi(x_i)\right)-\left(f(x_i) - \phi_*(x_i)\right)$ 应当与 $\left(f(x_i) - \phi(x_i)\right)$ 符号相同. 说明 $\phi_* (x_i) -  \phi (x_i)$ 存在 $(n+2)$ 个点的交错点组以及有 $n+1$ 个零点, 必然为零多项式. 但这又与假设中的严格不等号矛盾.

#### Chebyshev 等振荡定理
Chebyshev Equioscillation 定理: 设 $f(x)\in C[a,b],\ \phi(x)\in \mathbb R_n[x]$. 那么 $f(x) - \phi(x)$ 存在至少 $n+2$ 个点的 Chebyshev 交错点组 等价于 $\phi(x)$ 是最佳一致逼近.

证: $\Leftarrow$: 若 $\phi(x)$ 是 $f(x)$ 的最佳一致逼近, 简记 $r=\Vert f(x) - \phi(x)\Vert_{\infty}$, 且只用考虑非平凡的情形 $r>0$. 设 $x_1$ 为 $[a,b]$ 上最小的满足 $|f(x_1) - \phi(x_1)| = r$ 的点, 不妨设 $f(x_1) - \phi(x_1) = r>0$, 再取区间 $x_2$ 为 $[x_1,b]$ 上最小的满足 $f(x_2) - \phi(x_2) = -r $ 的点, $x_3$ 为 $[x_2,b]$ 上最小的满足 $f(x_3) - \phi(x_3) = r $ 的点, ... 如此贪心得到 Chebyshev 交错点组 $x_1< x_2 < ... < x_m$ 下证 $m\geqslant n+2$. 

反证, 若 $m\leqslant n+1$. 设 $t_i$ 为区间 $[x_i,x_{i+1}]$ 上最大的满足 $f(t_i) - \phi(t_i) = 0$ 的点,
$x_1<t_1<x_2<t_2< ... < t_{m-1} < x_m$. 于是 $f(x)-\phi(x)$ 在 $(t_i,t_{i+1})$ 与 $(-1)^i$ 同号. 构造 $g(x) = \alpha (-1)^{m-1} \prod_{i=1}^{m-1} (x - t_i)$ 是不超过 $n$ 次多项式. 令 $a>0$, 则 $g(x)$ 在 $(t_i,t_{i+1})$ 与 $(-1)^i$ 同号. 那么, 当 $\alpha$ 是充分小的正数, 在每个区间 $(t_i,t_{i+1})$ 上 $f(x)-\phi(x) - g(x)$ 比 $f(x)-\phi(x)$ 更接近 $0$. 故 
$$\Vert f(x)-(\phi(x)+g(x))\Vert_\infty <
\Vert f(x) - \phi(x)\Vert_\infty$$
这与 $\phi(x)$ 作为最佳逼近矛盾, 因此 $m\geqslant n+2$.

$\Rightarrow$: 设 Chebyshev 交错点组为 $x_1,\dotsc,x_{n+2}$. 不妨 $f(x_1) - \phi(x_1)>0$. 若存在 $h(x)=\phi(x)+g(x)$ 优于 $\phi(x)$, 则需要 $g(x_1)>0,\ g(x_2)<0,\ \dotsc,\ g(x_{n+2})(-1)^{n+2}<0$. 说明 $g(x)$ 不为零多项式却有 $n+1$ 个零点, 矛盾.
