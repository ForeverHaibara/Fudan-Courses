# 4 约束满足问题

例如寻找八皇后的解是约束满足问题 (CSP, constraint satisfication problem). 例如, 若设第 $1\sim 8$ 行的皇后所在列的编号为 $x_i$. 则状态空间总共有 $8^8=16777216$ 种可能, 约束有 $C_{8}^2=28$ 条.

## 约束图

### 二元约束图

若每个约束只涉及两个变量, 则可以把每个约束视为边 (edge), 涉及的两个变量为端点 (vertex), 形成一张约束图.

### 多元约束图

若某些约束涉及超过两个变量, 可以把它将其强行转换为二元约束. 例如升维.

## 搜索算法

### 回溯搜素

即使用 DFS 对每个变量依次赋值搜索.


#### 排序

可以优先尝试可能性最少 或 带来限制最小的 变量, 可能可以快速达到目标.

#### 过滤

需要及时检查不满足约束的方法. 可以采用:
* 前向检查: 每次准备新赋值的时候检查一下是否已经产生冲突
* 约束传播: 根据已经确定的变量增加约束
* 弧相容: 根据约束变量的连接进行检查或赋值

### 树结构

若约束图是一棵树 (没有环), 则非常快: 按拓扑顺序依次赋值即可, 是多项式时间复杂度. 对于一般的, 可以考虑先赋值一些变量使得剩余的约束图形成一棵树.

### 爬山法

先对所有变量任意赋值, 然后每次随机选一个不满足约束的变量, 将其赋值为冲突最少的新值. 逐步修改.
