# 10 几何

## 线性空间

#### 线性空间

又称向量空间, 定义: $V$ 为一个非空集合, $\mathbb P$ 为域, 定义加法 $+:V\otimes V\rightarrow V$, 数乘 $\cdot:\mathbb P\otimes V\rightarrow V$, 且满足以下 9 条定律:

1. 加法交换律: $a+b=b+a$
2. 加法结合律: $(a+b)+c=a+(b+c)$
3. 零元: $a+0=a$ 且 $0\in V$
4. 逆元: $a+(-a)=0$ 且 $-a\in V$
5. 幺元: $1a = a$ 且 $1\in P$
6. 数乘结合律: $(\lambda \mu) a = \lambda (\mu a)$
7. 分配律1: $\lambda (a+b) = \lambda a+\lambda b$
8. 分配律2: $(\lambda +\mu)a = \lambda a + \mu a$

构成一个 $V$ 在域 $\mathbb P$ 上的一个线性空间.

#### 空间和

$\mathcal V_1,\mathcal V_2$ 均为相同运算下, 域 $\mathbb P$ 上的线性空间, 定义其和为 
$$\mathcal V_1+\mathcal V_2 = \{a+b|\ a\in \mathcal V_1,\ b\in \mathcal V_2\}$$

则 $\mathcal V_1+\mathcal V_2$ 也是相同运算下, 域 $\mathbb P$ 上的线性空间.

#### 空间交

$\mathcal V_1,\mathcal V_2$ 均为相同运算下, 域 $\mathbb P$ 上的线性空间, 定义其和为 
$$\mathcal V_1\cap \mathcal V_2 = \{a|\ a\in \mathcal V_1,\ a\in \mathcal V_2\}$$

则 $\mathcal V_1\cap\mathcal V_2$ 也是相同运算下, 域 $\mathbb P$ 上的线性空间.

## 基

### 基与维数

对于有限维线性空间, 任何一个极大无关组都称为基. 这样的基总包含定值个元素, 该定值称为线性空间的维数. 任何一个非极大无关组都可以往其中继续添加元素使得最终构成一组基 (基扩充定理).

#### 维数公式

$$\dim \mathcal V_1 + \dim \mathcal V_2 = \dim (\mathcal V_1 +\mathcal V_2) + \dim (\mathcal V_1 \cap\mathcal V_2)$$

#### 定理

无限域上的有限维线性空间不能被有限个真子空间覆盖.

证: 任取一组基 $e_1,e_2,\dotsc,e_n$, 考虑元素 $a_k = k^1e_1 + k^2e_2+\dotsc + k^ne_k$, 其中 $k=1,2,\dotsc$

由范德蒙行列式知任意 $n$ 个 $a_k$ 线性无关. 故每一个真子空间至多覆盖 $n-1$ 个 $a_k$ (否则就存在大小为 $n$ 的线性无关组). 因此必有 $a_k$ 不被任何一个真子空间覆盖.

### 线性映射


映射 $f$ 将线性空间 $\mathcal V_1$ 中每个元素映到 $\mathcal V_2$ 中, 且 $\mathcal V_1$ 中的加法与数乘映射后在 $\mathcal V_2$ 仍然成立, 则 $f$ 为线性映射 (linear transformation).

$$f(a)+f(b)=f(a+b)\\ \lambda f(a)= f(\lambda a)$$

#### 线性变换

线性映射中 $\mathcal V_1 = \mathcal V_2$ 时称为线性变换.

#### 表示矩阵

取定线性空间 $\mathcal V_1$ 与 $\mathcal V_2$ 两组基 $E,F$ 则 $f(E)$ 在 $F$ 的坐标 $A$ 称为表示矩阵. 即
$$f(E) = FA$$

特别地, 若 $f$ 是线性变换, 一般取两组基相同, $E=F$.

#### 过渡矩阵

在表示矩阵的概念中, 有时需要解决问题: 已知一个坐标系下一点的坐标, 求其在另一个坐标系的坐标. 此时 $\mathcal V_1 = \mathcal V_2$, $f$ 为恒等映射, $E = FA$, 称 $A$ 为 $E$ 到 $F$ 的过渡矩阵.


#### 像空间

即线性映射的值域, 又称作列空间 (column space), 又记作 ${\rm Range}(f)$, ${\rm col}(f)$. 亦可用 $A$ 代替 $f$ 表示.
$${\rm Im}(f) = \{f(x)|x\in\mathcal V_1\}$$

#### 核空间

即线性映射的零点, 又称作零空间 (nullspace), 又记作 $\mathcal N(f)$. 亦可用 $A$ 代替 $f$ 表示.

$${\rm Ker}(f) = \{x|f(x)=0, \ x\in\mathcal V_1\}$$

#### 秩与零度

$$ {\rm rank}(f)=\dim {\rm Im}(f)\\ {\rm Null}(f)=\dim {\rm Ker}(f)$$

#### 正交补

若全空间可分解为两个不相交的子空间之和: $\mathcal V=\mathcal V_1 + \mathcal V_2$ 且 $\mathcal V_1\cap \mathcal V_2=\emptyset$, 且 $\forall \alpha\in \mathcal V_1,\ \beta\in \mathcal V_2,\  \alpha\perp \beta$, 则称 $\mathcal V_1,\mathcal V_2$ 互为正交补 (orthogonal complement).

对于任意矩阵 $A\in \mathbb K^{m\times n}$, ${\rm Im}(A)$ 与 ${\rm Ker}(A^T)$ 总互为正交补, 
${\rm rank}(f)+ {\rm Null}(f)=n$.

## 内积空间

线性空间 $\mathcal V$ 是域 $\mathbb C$ 上的线性空间, 二元函数 $\langle \cdot,\cdot\rangle : \mathcal V\times \mathcal V\rightarrow \mathbb C$ 满足:

1. 正定性: $\langle x,x\rangle \geqslant 0$ 且取等当且仅当 $x=0$. 
2. (复)对称性: $\langle x,y\rangle = \overline{\langle y,x\rangle}$.
3. 线性性: $\langle x_1+x_2,y\rangle = \langle x_1,y\rangle +\langle x_2,y\rangle$.
4. 齐次性: $\langle \alpha x,y\rangle = \alpha \langle x,y\rangle$. 

注意, $\langle  x,\alpha y\rangle = \overline \alpha \langle x,y\rangle$. 

则称该线性空间与该内积构成一个内积空间. 二元函数 $\langle \cdot,\cdot \rangle$ 称为内积 (inner product). 有限维复内积空间也称为酉空间 (unitary space). 有限维实内积空间称为欧式空间 (Euclid space).

最常用的内积空间: $\langle x,y\rangle =y^*x$ 是内积. 当空间是实空间时, $\langle y,x\rangle = y^Tx=x^Ty$ 是内积.

### 范数

定义内积诱导的范数 $\Vert x\Vert = \sqrt{\langle x,x\rangle}$. 下证明这是范数.

证:

1. 正定性: $\Vert x\Vert\geqslant 0$ 且取等当且仅当 $x=0$, 由内积的正定性显然.
2. 齐次性: 
$$\Vert \alpha x\Vert = \sqrt{\langle \alpha x,\alpha x\rangle}=\sqrt{\alpha \langle x,\alpha x\rangle}=\sqrt{\alpha \overline{\langle \alpha x,x\rangle}}=\sqrt{\alpha \overline{\alpha \langle x,x\rangle}}=|\alpha|\sqrt{\overline{\langle x,x\rangle}}=|\alpha|\sqrt{\langle x,x\rangle}=|\alpha|\Vert x\Vert$$

3. 三角不等式: 令 $t\in \mathbb R$, 

$$\Vert x+ty\Vert^2 = \langle x+ty,x+ty\rangle = \Vert x\Vert^2 + |t|^2\Vert y\Vert^2 + 2t\Re \langle y,x\rangle \geqslant 0$$

从而 $\Delta = 4\left(\Re \langle y,x\rangle \right)^2-4\Vert x\Vert^2\Vert y\Vert^2\leqslant 0$

特别地, 取 $t=1$, 并代入上式得

$$\Vert x+y\Vert^2 \leqslant  \Vert x\Vert^2 + \Vert y\Vert^2 + 2\sqrt{\Vert x\Vert^2\Vert y\Vert^2} = 
\left(\Vert x\Vert + \Vert y\Vert \right)^2$$

#### 柯西不等式

$$\Vert x\Vert\Vert y\Vert\geqslant |\langle x,y\rangle| $$

证: 在三角不等式成立性的证明中, 已经证明 $\Vert x\Vert\Vert y\Vert\geqslant |\Re \langle y,x\rangle|$. 若对于实内积空间已经得证, 而对于复内积空间用 $zy$ 代 $y$, 其中 $|z|=1$ 为单位复数, 那么
$$ \Vert x\Vert\Vert y\Vert= \Vert x\Vert\Vert zy\Vert 
\geqslant |\Re \langle zy,x\rangle| = |\Re (z\langle y,x\rangle)|$$

特别地, 取单位复数 $z$ 使得 $z\langle y,x\rangle\in \mathbb R$ 即得证.

#### 极化恒等式

对于复内积空间,

$$\langle x,y\rangle = \frac{1}{4}\left(\Vert x+y\Vert^2 - \Vert x-y\Vert^2 +i\Vert x+iy\Vert^2 -i\Vert x- iy\Vert^2 \right)$$

特别地, 对于实内积空间,
$$\langle x,y\rangle = \frac{1}{2}\left(\Vert x+y\Vert^2 - \Vert x-y\Vert^2 \right)$$

#### 平行四边形等式

$$\Vert a+b\Vert^2 + \Vert a-b\Vert^2 = 2\Vert a\Vert^2 + 2\Vert b\Vert^2$$

#### 余弦定理

$$\Vert a+b\Vert^2 = \Vert a\Vert^2 +\Vert b\Vert^2 + 2\Re \langle a,b\rangle$$

#### 正交

若 $a,b$ 满足 $\langle a,b\rangle = 0$ 则称 $a,b$ 正交 (orthogonal) 或垂直 (perpendicular).

### Gram 矩阵

若 $E=[e_1,\dotsc,e_n]$ 为一组向量, 则称其构成的 度量矩阵 / Gram 矩阵 (Gram matrix) 为 $G=[\langle e_i,e_j\rangle]_{1\leqslant i,j\leqslant n}$.

特别地, 若内积被定义为通常的 $\langle e_i,e_j\rangle = e_j^*e_i$, 则 $G = [e_j^*e_i]_{1\leqslant i,j\leqslant n} = E^*E$.

Gram 矩阵是 Hermite 半正定矩阵. 特别地, 若这组向量是一组基且 Gram 矩阵是单位阵, 则称这组基在该内积空间是标准正交基.

#### Cauchy 矩阵

Cauchy 矩阵, 又称 Hilbert 矩阵, 是形如 $H_n = [\frac{1}{i+j-1}]_{1\leqslant i,j\leqslant n}$ 的半正定矩阵. 

若在实函数空间$C[0,1]$ 上定义 (实) 内积 
$$\langle f,g\rangle = \int_0^1 f(t)g(t)dt$$

容易验证其满足内积定义且若取 $E=[t,t^2,\dotsc,t^n]$, 则容易验证其 Gram 矩阵正是 Hilbert 矩阵 $G_E = H_n$.

## 仿射空间

设 $\mathcal U,\mathcal V$ 是域 $\mathbb P$ 上两个线性空间, 且映射 $f:\ \mathcal U\rightarrow \mathcal V$ 满足性质: 任意 $x,y\in U, \forall \alpha \in \mathbb P$, 
$$f(\alpha x+(1-\alpha) y) = \alpha f(x)+(1-\alpha) f(y)$$

则称 $f$ 是仿射 (affine transformation). 

仿射 $f(x)$ 可表示为 $f(x) = f(0) + g(x)$, 其中 $g(x)$ 是线性映射.
