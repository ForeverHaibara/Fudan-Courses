# 3 矩阵范数

## 向量范数

#### 定义

在 $\mathbb C^n$ (或 $\mathbb R^n$) 上满足下列三条性质的函数 $\Vert x\Vert$ 称为范数 (norm).

(1) 正定性: $\Vert x\Vert \geqslant 0$, 且等号成立当且仅当 $x=0$ .

(2) 齐次性: $\Vert kx\Vert = |k| \Vert x\Vert$ .

(3) 三角不等式: $\Vert x+y\Vert \leqslant \Vert x\Vert + \Vert y\Vert$ .

#### 有限维范数

对于一种有限维空间的范数, 它满足下列三条性质:

(1) (连续性) $\Vert x\Vert$ 是关于 $x$ 坐标的连续函数.

(2) 依范数收敛等价于依坐标收敛.

(3) (等价性) 任何两种范数 $\Vert \cdot \Vert_a$, $\Vert \cdot \Vert_b$, 存在正数 $m,M$ 使得对于一切非零的 $x$ 有:

$$m\leqslant \frac{\Vert x\Vert_a}{\Vert x\Vert_b} \leqslant M$$

这说明任何两种范数趋于无穷的量级是相同的.

#### p-范数

p-范数是范数中的一类, 也称为 $l^p$-范数, 是 Minkowski 距离下的范数, 设 $x$ 的坐标为 $(x_1,\dotsc,x_n)^T$, 则

$$\Vert x\Vert_p = \left(\sum_{k=1}^n |x_k|^p\right)^\frac 1 p\quad\quad (1\leqslant p\leqslant +\infty)$$

可以验证它满足范数的三条定义, 它具有如下性质:

(1) 对于固定的 $x$, $\Vert x\Vert_p$ 关于 $p$ 连续且单调不增.

(2) Holder 不等式: 若 $\frac{1}{p}+\frac{1}{q} = 1$, 则 $|y^*x|\leqslant \Vert x\Vert_p\Vert y\Vert_q$ .

(3) Minkowski 不等式 (三角不等式): $\Vert x+y\Vert_p\leqslant \Vert x\Vert_p + \Vert y\Vert_p$ .

有以下常用范数:

1. 1范数: $\Vert x\Vert_1 = \sum_{k=1}^n |x_k|$ .
2. 2范数: $\Vert x\Vert_2 = \sqrt{x^*x}$ .
3. $\infty$ 范数: $\Vert x\Vert_{\infty} = \max_k |x_k|$ .



## 矩阵范数

#### 相容范数

若某矩阵-值函数 $\Vert A\Vert$ 满足如下四条性质, 则称为相容范数. 相容范数是满足相容性的矩阵范数. 在无特别说明的情况下, 我们此处只讨论满足相容性的矩阵范数.

(1) 正定性: $\Vert A\Vert \geqslant 0$, 且等号成立当且仅当 $A=0$ .

(2) 齐次性: $\Vert kA\Vert = |k| \Vert A\Vert$ .

(3) 三角不等式: $\Vert A+B\Vert \leqslant \Vert A\Vert + \Vert B\Vert$ .

(4) 相容性: 对于任何 $A\in \mathbb C^{m\times n}, B\in\mathbb C^{n\times k}$, $\Vert AB\Vert \leqslant \Vert A\Vert  \Vert B\Vert$ .

由相容性, 相容范数显然使得单位阵范数不小于 $1$ :  $\Vert I_n\Vert\geqslant 1$.
#### 诱导范数

诱导范数 (induced norm), 亦称算子范数, 是一类相容的矩阵范数: 若 $\Vert x\Vert$ 是向量范数, 则定义该范数诱导的矩阵范数:

$$\Vert A\Vert = \sup_{x\neq 0}\frac{\Vert Ax\Vert}{\Vert x\Vert} = \sup_{\Vert x\Vert = 1}\Vert Ax\Vert$$

可以验证这的确满足矩阵范数的四条性质: 正定性与齐次性显然. 

三角不等式:

$$ \Vert A+B\Vert =  \sup_{\Vert x\Vert = 1}\Vert Ax+Bx\Vert
\leqslant \sup_{\Vert x\Vert = 1}\left(\Vert Ax\Vert + \Vert Bx\Vert\right)
\leqslant \sup_{\Vert x\Vert = 1}\Vert Ax\Vert + \sup_{\Vert x\Vert = 1}\Vert Bx\Vert
=\Vert A\Vert + \Vert B\Vert $$

相容性:

$$\Vert AB\Vert= \sup_{x\neq 0}\frac{\Vert ABx\Vert}{\Vert x\Vert} 
= \sup_{Bx\neq 0}\frac{\Vert ABx\Vert}{\Vert Bx\Vert} \cdot \frac{\Vert Bx\Vert}{\Vert x\Vert}
\leqslant \sup_{Bx\neq 0}\frac{\Vert ABx\Vert}{\Vert Bx\Vert} \cdot \sup_{x\neq 0}\frac{\Vert Bx\Vert}{\Vert x\Vert}
\leqslant \Vert A\Vert \Vert B\Vert$$

且显然满足单位阵的诱导范数为 $1$.

#### p-范数

当诱导范数中选取的向量范数是向量的 p-范数时即为矩阵的 p-范数. 常用的 p-范数: ([2] pp. 344)

1. 1范数 (列和范数): $\Vert A\Vert_1 = \max_{j}\sum_{i=1}^m |a_{ij}|$.
2. 2范数 (谱范数): $\Vert A\Vert_2 = \sqrt{\rho (A^*A)} = \max \sigma(A)$.
3. $\infty$ 范数 (行和范数): $\Vert A\Vert_{\infty} = \max_{i}\sum_{j=1}^n |a_{ij}|$.

#### 2-范数

2范数 (谱范数) 具有如下性质:

(1) $\Vert A\Vert_2 = \sqrt{\rho (A^*A)} = \max \sigma(A)$, 其中 $\rho (A^*A)$ 表示 $A^*A$ 的最大特征值, $\sigma(A)$ 为其奇异值.

证: $$\Vert A\Vert_2 = \sup_{\Vert x\Vert = 1}\Vert Ax\Vert
=\sup_{\Vert x\Vert = 1} \sqrt{x^* A^*Ax} = \sqrt{\rho (A^*A)}$$

(2) $\Vert A\Vert_2^2\leqslant \Vert A\Vert_1\Vert A\Vert_{\infty}$ .

证: 注意 $\Vert A^*Ax\Vert_2 \leqslant \Vert A^*Ax\Vert_1$, 故
$$\Vert A\Vert_2^2=\Vert A^*A\Vert_2=\sup_{\Vert x\Vert = 1}\Vert A^*Ax\Vert_2
\leqslant \sup_{\Vert x\Vert = 1}\Vert A^*Ax\Vert_1 = \Vert A^*A\Vert_1
\leqslant \Vert A\Vert_1\Vert A^*\Vert_{1} = \Vert A\Vert_1\Vert A\Vert_{\infty}$$

(3) 酉不变性: 任意酉矩阵 $U$,$V$, $\Vert U^*AV\Vert_2 = \Vert A\Vert_2$. (由(1)显然)

#### F-范数

F-范数又称 Frobenius 范数, 是矩阵中所有元素的平方和开根. 可以验证它是相容范数但不是 p-范数.

$$\Vert A\Vert_F = \sqrt{\sum_{ij}|a_{ij}|^2}$$

Frobenius 范数具有如下性质:

(1) $\Vert A\Vert_F  = {\rm tr}(A^*A) = {\rm tr}(AA^*) = \sqrt{\sum \sigma^2(A)}$

(2) 与2范数的相容: 
$$\Vert AB\Vert_F \leqslant \min\{\Vert A\Vert_2\Vert B\Vert_F,\Vert A\Vert_F\Vert B\Vert_2\}
\leqslant \max\{\Vert A\Vert_2\Vert B\Vert_F,\Vert A\Vert_F\Vert B\Vert_2\}
\leqslant \Vert A\Vert_F\Vert B\Vert_F $$

(3) 酉不变性: 任意酉矩阵 $U$,$V$, $\Vert U^*AV\Vert_F = \Vert A\Vert_F$. (由(1)显然)

### 谱半径

若 $A$ 是方阵, 则记其特征值的最大模长为 $\rho (A)$, 称为谱半径 (spectral radius). 谱半径不是范数 (显然不满足三角不等式). 谱半径具有如下性质: ([2] pp. 347)

(1) $A$ 是方阵, 则任意相容范数 $\Vert A\Vert\geqslant \rho (A)$.

证: 取 $A$ 的模长最大的特征值 $\lambda$ 与特征向量 $x_0$, 则
$$\Vert A\Vert \Vert x\Vert \geqslant \Vert Ax\Vert = \Vert \lambda x\Vert = |\lambda| \Vert x\Vert$$

(2) 对于一个方阵 $A$, $\forall \epsilon>0$, 存在一种矩阵范数使得 $\rho(A)\leqslant \Vert A\Vert < \rho(A) + \epsilon$ .  ([2] pp. 347) 

(3) (Gelfand) 对于一个方阵 $A$, 则任意一种范数 (不必相容) 满足 $\rho (A) = \lim_{k\rightarrow \infty} \Vert A^k\Vert ^\frac 1 k$ .

(4) 若 $\rho (A)<1$, 则 $\lim_{k\rightarrow \infty} A^k= 0$. 若 $\rho(A)>1$, 则 $ A^k$ 无界.

#### Neumann 级数

对于方阵 $A$, 当且仅当 $\rho(A)<1$ 时 $\sum_{k=0}^{\infty}A^k$ 收敛, 且
$$\sum_{k=0}^{\infty} A^k = \frac{1}{I - A} = (I-A)^{-1}$$
