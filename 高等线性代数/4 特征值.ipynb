# 6 特征值

任何方阵 $A\in \mathbb C^{n\times n}$, 若存在非零列向量 $x$ 与 常数 $\lambda$ 使得 $Ax = x\lambda$, 则称 $\lambda$ 为 $A$ 的一个特征值, $x$ 为 $\lambda$ 对应的一个特征向量 (eigenvalue). 若 $x$ 是特征向量 (eigenvector), 则可以通过 $\lambda = \frac{x^*Ax}{x^*x}$ 求出对应的特征值.

#### 特征多项式

方阵 $A$ 有 $n$ 个特征值, 它们是 $n$ 次多项式 $p(\lambda) = {\rm det}(\lambda I - A) = 0$ 的所有复根. 该首一 $n$ 次多项式称为 $A$ 的特征多项式 (characteristic polynomial).

#### 韦达定理

对特征多项式使用韦达(Vieta)定理, 知 $\sum_{k=1}^n \lambda_k = {\rm tr}(A)$, $\prod_{k=1}^n \lambda_k = {\rm det}(A)$ .

#### 重数

对于方阵 $A$ 的一个特征值 $\lambda$, 特征多项式 $p(\lambda) = 0$ 的根 $\lambda$ 的重数称为代数重数 (algebraic multiplicity). $(\lambda I - A)x = 0$ 的解空间的度数 (零度) 被称为几何重数 (geometric multiplicity). 

对于任何特征值 $\lambda$, 其代数重数不小于几何重数. 

所有特征值的代数重数之和为 $n$. 

#### 谱映射定理

$f(A)$ 是矩阵多项式. 若 $A$ 的 $n$ 个特征值为 $\lambda_1,\dotsc,\lambda_n$, 则 $f(A)$ 的 $n$ 个特征值为 $f(\lambda_1),\dotsc, f(\lambda_n)$. ([1] p. 268)

#### 相似变换

对于方阵 $A$, 任取一个可逆矩阵 $P$, 变换 $A\mapsto P^{-1}AP$ 称为相似变换. 若存在相似变换使得 $A\mapsto B$, 则称 $A,B$ 相似, 记作 $A\sim B$.

相似变换具有如下性质:

(1) 自反性: $A\sim A$.

(2) 对称性: $A\sim B \Leftrightarrow B\sim A$. 

(3) 传递性: $A\sim B, B\sim C \Rightarrow A\sim C$.

(4) 相似变换不改变矩阵特征值, 特征多项式, 特征值的代数重数与几何重数. ([1] p. 264)

(5) 相似变换与域无关: 若实矩阵 $A,B$ 复相似, 则它们也实相似.

证: 利用 $\lambda$-矩阵 .

## Jordan 标准型

#### Jordan 块

形如 $$J_k(\lambda) = \left[\begin{matrix} \lambda & 1 & & &  \\
 & \lambda & 1 & & \\
 & & \ddots & \ddots & \\
 & & &  \lambda & 1\\
 & & & & \lambda\end{matrix}\right]$$ 的$k\times k$矩阵称为约尔当块 (Jordan block). ([1] p. 331) 其特征值是 $k$ 重的 $\lambda$. 


#### Jordan 标准型

对角元为若干个 Jordan 块的分块对角方阵称为约尔当标准型 (Jordan canonical form), 即
$$J=\left[\begin{matrix}J_{d_1}(\lambda_1)& & & \\   
 & J_{d_2}(\lambda_2) & & \\
 & & \ddots & \\
 & & & J_{d_m}(\lambda_m)\end{matrix}\right]$$

每个方阵都相似于一个 Jordan 标准型.

<font color=red>若不计 Jordan 标准型的 Jordan 块的次序, 则两个方阵相似当且仅当它们的 Jordan 标准型相同.</font>

推论: $A$ 与 $A^T$ 相似.

#### 重数

$A$ 的特征值 $\lambda$ 的代数重数等于其 Jordan 标准型中, 属于特征值 $\lambda$ 的 Jordan 块维数之和. 

$A$ 的特征值 $\lambda$ 的几何重数等于属于 $\lambda$ 的 Jordan 块的个数.


#### 初等因子

若矩阵 $A$ 的 Jordan 标准型如下所示, 则其初等因子组为 $(\lambda - \lambda_1)^{d_1},\dotsc, (\lambda - \lambda_m)^{d_m}$.

 两个方阵相似当且仅当初等因子组 (不计次序) 相同.
([1] p. 314)

$$J=\left[\begin{matrix}J_{d_1}(\lambda_1)& & & \\   
 & J_{d_2}(\lambda_2) & & \\
 & & \ddots & \\
 & & & J_{d_m}(\lambda_m)\end{matrix}\right]$$


#### 对角化

一个方阵可以相似于对角阵则称其可以对角化 (diagonalize), 以下是可以对角化的充要条件:

(1) Jordan 标准型是对角阵, 即只有大小为 1 的 Jordan 块.

(2) 所有特征值代数重数 = 几何重数.

特别地, 如果方阵没有重特征值, 则必定可以对角化.

## Cayley-Hamilton 定理

若 $A$ 的特征多项式是 $p(\lambda)$, 则 $p(A) = 0$. 

#### Jordan 块

$J_k(\lambda)$ 的极小多项式是 $p(J_k(\lambda)) = (x - \lambda)^k$. 显然 $(J_k(\lambda) - \lambda I)^k = O$

#### 极小多项式

使得 $f(A) = 0$ 的次数最低的首一多项式 (不是零多项式) 称为 $A$ 的极小多项式 (minimal polynomial). 

对于任意多项式满足 $g(A)=0$, 则极小多项式必定是 $g$ 的因式.
 
由 Cayley-Hamilton 定理只要 $A\neq 0$, 极小多项式次数不超过 $n$. 

若矩阵 $A$ 的互不相同的特征值为 $\lambda_1,\dotsc ,\lambda_m$, 且每个特征值对应的最大 Jordan 块的维数为 $d_1,\dotsc, d_m$, 则
$A$ 的极小多项式为 $$f(\lambda) = (\lambda - \lambda_1)^{d_1}\dotsc (\lambda-\lambda_m)^{d_m}$$

### Frobenius 矩阵

Frobenius 矩阵 (Frobenius companion matrix) 是如下矩阵: ([1] p. 304)
$$\left[\begin{matrix}
0 & & & & -a_0\\
1 &0& & & -a_1\\
  &1&0& & -a_2\\
  & &\ddots &\ddots & \vdots\\
  & & &1& -a_{n-1}\end{matrix}\right]$$

它的特征多项式兼极小多项式是 $$f(\lambda) = \lambda^n + a_{n-1}\lambda^{n-1}+\dotsc + a_1\lambda + a_0$$

### Taylor 公式

若 $f(x)$ 在 $x=\lambda$ 展开式为 $f(x) = f(\lambda)+(x-\lambda)f'(\lambda)+\frac{(x-\lambda)^2f''(\lambda)}{2!} +\dotsc$ , 则其作用在 $\lambda$ 的 Jordan 块结果为:
$$f\left(J_k(\lambda)\right) = \left[\begin{matrix} f(\lambda) & f'(\lambda) & \frac{f''(\lambda)}{2!}&\ddots &\frac{f^{(n-1)}(\lambda)}{n!}  \\
 &  f(\lambda) & f'(\lambda) & \ddots&\ddots \\
 & & \ddots & \ddots & \frac{f''(\lambda)}{2!}\\
 & & &   f(\lambda)& f'(\lambda)\\
 & & & &  f(\lambda)\end{matrix}\right]$$

### Nonderogatory

### Nonderogatory

若矩阵 $A$ 所有特征值的几何重数为 1, 则称该矩阵是 nonderogatory 的. ([2] p. 178)

#### 交换性

若 $A$ 是 nonderogatory 的, 且 $AB=BA$, 则 $B$ 一定可表为 $A$ 的多项式. ([2] p. 178)

### Jordan–Chevalley 分解

任何方阵 $A$, 存在可对角化矩阵 $S$ 与幂零矩阵 $N$ 使得 $A=S+N$ 且 $SN=NS$.

证: 将 Jordan 标准型对角部分与 "次对角线" 分离出来即可.

### Lie's Theorem 李定理

$A_1,\dotsc,A_n$ 两两可交换 (commutative), 则它们可以同时酉上三角化, 即存在酉矩阵 $U$ 使得所有 $U^*A_iU$ 为上三角.
