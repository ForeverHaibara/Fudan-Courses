# 1 Probability and Random Variables


* Random Experiment : An experiment whose outcome cannot be determined in advance, but is nnevertheless still subeject to analysis. E.g. tossing a die / measuring the amount of rainfall in Brisbane in January.

* Probability Space: The set $\Omega$ that contains all possible outcomes of a random experiment. It is also called the outcome space (in probability theory) or the sample space (in statistical theory). E.g. tossing a die has probability space $\Omega = \{1,2,3,4,5,6\}$.

## Measurable Space

### $\sigma$-Algebra

Let $\mathcal F$ be a collection of subsets of a sample space. It is called a $\sigma$-field or a $\sigma$-algebra if and only if it has the following three properties. 
* $\emptyset\in \mathcal F$ 
* $A\in \mathcal F\Rightarrow A^C \in \mathcal F$
* $A_i\in \mathcal F\ (i=1,2,\dotsc)\Rightarrow \bigcup A_i\in \mathcal F$ (can be countably infinitely many)

As a consequence, $\bigcap A_i = \left(\bigcup A_i^C\right)^C\in \mathcal F$ and also $A\setminus B = A\cap  B^C\in \mathcal F$.

We denote the smallest $\sigma$-field containing a collection-of-set $\mathcal C$ as $\sigma(\mathcal C)$.

E.g. When $\mathcal C$ contains only one nonempty subset, e.g. $\mathcal C = \{A\}$, then $\sigma(\mathcal C) = \{\emptyset, A,A^C,\Omega\}$.
  
### Borel $\sigma$-field 

On $\Omega = \mathbb R$, let $\mathcal C$ be the collection of all finite open intervals on $\mathbb R$, then $\mathcal B=\sigma(\mathcal C)$ is named the Borel $\sigma$-field. We will keep using the notation $\mathcal B$ in the following context.


### Measure 
Let $\mathcal F$ be a $\sigma$-algebra on $\Omega$. A set function $\nu$ defined on $(\Omega, \mathcal F)$ is called a measure if and only if it has following three properties,
* $\forall A\in \mathcal F$, $0\leqslant \nu(A)\leqslant +\infty$
* $\nu(\emptyset) = 0$
* if $A_i\in \mathcal F\ (i=1,2,\dotsc)  $ are disjoint, then $\nu(\cup A_i)=\sum \nu(A_i)$  (can be countably infinitely many)

### Properties 
1. Monotonicity: If $A\subset B$, then $\nu(A)\leqslant \nu(B)$
   
Proof: $\nu(B) = \nu(A)+\nu(B\setminus A)\geqslant \nu(A)$. Here we have known  $B\setminus A = B\cap A^C\in \Omega$.

2. Subadditivity: If $A\subset  \bigcup_{m=1}^{\infty}A_m$, then $\nu(A)\leqslant \sum_{m=1}^{\infty} \nu(A_m)$.

Proof: 
$$\nu(A) = \nu\left(\bigcup_{m=1}^\infty A_m\right) = \nu\left(\bigcup_{m=1}^\infty \left(A_m\cap \left(\cup_{j=1}^{m-1} A_j\right)^C\right)\right)=\sum_{m=1}^\infty \nu \left(A_m\cap \left(\cup_{j=1}^{m-1} A_j\right)^C\right) \leqslant \sum_{m=1}^\infty \nu (A_m).$$

3. Continuity: If $A_i\uparrow A$, i.e. $A_1\subset A_2\subset \dotsc$ and $\bigcup_{i=1}^\infty A_i = A$, then we have $\lim_n \nu(A_n) = \nu(A)$. Similarly, if $A_i\downarrow A$, i.e. $\dotsc \subset A_2\subset A_1$ and $\bigcap_{i=1}^\infty A_i =A$, then we have $\lim_n \nu (A_n) = \nu(A)$.

Proof: We only prove the first case where $A_i\uparrow A$. This is by 
$$\nu (A) = \nu\left(\bigcup_{n=1}^\infty (A_n\cap A_{n-1}^C)\right)
=\lim_N \sum_{n=1}^N \nu  (A_n\cap A_{n-1}^C) =\lim_N \nu (A_N).$$


### $\sigma$-finite Measure

Measure $\mu$ is said to be $\sigma$-finite if there exists a finite sequence of sets, $A_1,\dotsc,A_n\in \mathcal A$ such that $\mu(A_i)<\infty$ and $\bigcup_{i=1}^n A_i = \Omega$.

### Probability Measure 

To be more specific, if $\nu(\Omega)  =1$, then we call $(\Omega,\mathcal F,\nu )$ is a probability measure. In this case we often use notation $\mathbb P$ to refer to a probability measure rather $\nu$.

### Cumulative Distribution Function

If $(\mathbb R, \mathcal B,\mathbb P)$ is a probability measure on Borel $\sigma$-algebra, we define a function $F(x) = \mathbb P((-\infty,x])$ and call it the cumulative distribution funciton (CDF) of $\mathbb P$.

#### Properties

1. $F(-\infty) = \lim_{x\rightarrow -\infty} F(x)=0$
2. $F(+\infty)=\lim_{x\rightarrow +\infty} F(x)=1$
3. $F$ is non-decreasing.
4. $F$ is right continuous, i.e. $\lim_{y\rightarrow x_+}F(y) = F(x)$.

Proof: 1. and 2. are because $(-\infty, x]\downarrow \emptyset$ and $(-\infty, x]\uparrow \mathbb R$. 3. is simply given by the monotonicity of measure. 4. is because $(-\infty, y]\downarrow (-\infty, x]$ when $y\downarrow x$.

Note: $(-\infty, y]\uparrow (-\infty, x)\neq (-\infty, x]$ when $y\uparrow x<+\infty$, so $F$ might not be left continuous.

## Stieltjes Measure

**Theorem** We call $F$ as Stieltjes measure function if $F:\mathbb R\rightarrow \mathbb R$ is non-decreasing and is right-continuous. In this case, there exists a unique measure $\mu$ on $(\mathbb R, \mathcal B)$ such that 
$\mu((a,b]) = F(b)- F(a)$.

Stieltjes characterizes a sufficient condition for a function $F$ to be a cumulative function of a measure on $(\mathbb R,\mathcal B)$.

### (a) Relation Between Semialgebra and Algebra

First we introduce the concepts of semialgebra and algebra. 

#### Semialgebra 


A collection $\mathcal S$ of sets is said to be a semialgebra if (i) it is closed under intersection, i.e. $S,T\in \mathcal S\Rightarrow S\cap T\in \mathcal S$ and (ii) if $S\in \mathcal S$, then $S^C$ is a finite disjoint union of sets in $\mathcal S$.

E.g. $\mathcal S = \left\{\emptyset\right\}\cup \left\{(a,b]: -\infty\leqslant a\leqslant b \leqslant +\infty\right\}$ is a semi-algebra.

#### Algebra

A collection $\mathcal A$ of subsets of $\Omega$ is called an algebra (or field) if $A,B\in \mathcal A\Rightarrow A^C\in \mathcal A,\ A\cup B\in \mathcal A$.

<br>


**Theorem** If $\mathcal S$ is a semialgebra, then $\bar {\mathcal S} = \{\rm finite \ disjoint \ unions \ of \ \mathcal S\}$ is an algebra. We call $\bar{\mathcal S}$ the algebra generated from $\mathcal S$.

**Proof** Given arbitrary $A\in \bar{\mathcal S}$, it is finite disjoint of unions in $\mathcal S$. Say, $A = +_{i=1}^n A_i$ where $A_i\in \mathcal S$. Then, $A^C = \bigcap_{i=1}^n A_i^C$. Note that $A_i^C\in \mathcal S$ by the definition of semialgebra, we have $A^C$ being the intersection of finitely many sets in $\bar{\mathcal S}$ should also lie in $\bar{\mathcal S}$. Next, if given $B\in \bar{\mathcal S}$ and $B=+_{i=1}^m B_i$ where $B_i\in\mathcal S$, then $A\cap B = +_{i,j}A_i\cap B_j\in \bar{\mathcal S}$. It follows that $A\cup B = (A^C\cap B^C)^C\in \bar{\mathcal S}$.

#### Measure

Let $\mathcal A$ be an algebra, we call a set function $\mu$ defined on $\mathcal A$ a measure if and only if it has following two properties：
* $\forall A\in\mathcal A,\ 0\leqslant \mu(A)\leqslant +\infty$
* $\mu(\emptyset)=0$
* if $A_i\in\mathcal A$ are disjoint and $ \bigcup_{i=1}^\infty A_i\in \mathcal A$, then $\mu(\bigcup_{i=1}^\infty A_i) = \sum_{i=1}^\infty \mu(A_i)$.

Note that the only difference between $\sigma$-algebra and algebra is that the latter does not require closure under countably infinite disjoint union. Similarly, the measure of the algebra has additivity only when the disjoint union is in the algebra.

### (b) Carathéodory Extension Theorem

We introduce the Carathéodory extension theorem, it is the last sentence stated in the following theorem B. (The proof is too long and one had better first take a look at part (c).)

**Theorem B** Let $\mathcal S$ be a  semialgebra and let $\mu$ be a function defined on $\mathcal S$ such that $\mu(\emptyset) = 0$ and $\mu\geqslant 0$. Suppose 
(i) if $S\in \mathcal S$ is a finite disjoint union of sets $S=+_{i=1}^nS_i\in \mathcal S$, then $\mu(S) = \sum_{i=1}^n \mu(S_i)$ and (ii) if $S_i,S\in \mathcal S$ with $S = +_{i=1}^\infty S_i$, then $\mu(S)\leqslant \sum_{i=1}^\infty \mu(S_i)$. Then $\mu$ has a unique extension $\bar{\mu}$ that is a measure on $\bar{\mathcal S}$ the algebra generated by $\mathcal S$. If $\bar{\mu}$ is $\sigma$-finite then there is a unique extension $\nu$ that is a measure on $\sigma(\mathcal S)$.

To show this, we begin with a lemma.

<br>

**Lemma** Suppose that $\mu(\emptyset)=0,\ \mu\geqslant 0$ and, as long as $S\in \mathcal S$ is a finite disjoint union of $S_i\in \mathcal S$,  $\mu(S) = \sum \mu(S_i)$ holds. Then, if for any set $A\in   \bar {\mathcal S}$ with $A=+_{i=1}^n A_i,\ A_i\in \mathcal S$ we define $\bar \mu(A)=\sum_{i=1}^n \mu(A_i)$, we shall have

(1) if $A, B_i\in \bar{\mathcal S}$ with $A = +_{i=1}^n B_i$, then $\bar \mu (A) = \sum \bar \mu (B_i)$

(2) if $A,B_i\in \bar{\mathcal S}$ with $A\subset \cup_{i=1}^n B_i$, then $\bar{\mu}(A) \leqslant \sum \bar{\mu}(B_i)$.

**Proof** (1) is easy by decomposing each $B_i$ into $B_i=+_{j=1}^{k_i}C_{ij}$ where $C_{ij}\in \mathcal S$ are disjoint. So $\bar \mu(A) = \sum_{ij}\mu (C_{ij}) = \sum_{i=1}^n \bar\mu(B_i)$. (2) is by the trick that 
$$\bar \mu(A) = \bar\mu\left(\bigcup_{i=1}^n B_i\cap (\cup_{j=1}^{i-1}B_j)^C\right)
=\sum \bar \mu \left(B_i \cap (\cap_{j=1}^{i-1}B_j^C)^C\right)
\leqslant \sum \bar \mu \left(B_i \right).$$
The last inequality is beacuse $B_i\cap (\cap_{j=1}^{i-1}B_j^C)^C\in \mathcal S$ and $B_i\setminus (B_i\cap (\cap_{j=1}^{i-1}B_j^C)^C)\in \mathcal S$ plus the negativity / monotonicity of $\mu$.

Next we prove theorem B.

(b.1) Firstly, we will show that $\bar \mu$ is well-defined. That is, $\bar \mu(A)$ does not rely on its disjoint partition. Assume there are two different partitions of $A$, $A=+_{i=1}^n A_i = +_{j=1}^m A_j'$ where $A_i,A_j'\in\mathcal S$. Then, 
$$\sum_{i=1}^n \mu(  A_i) = \sum_{i=1}^n \mu(  +_{j=1}^m A_i\cap A_j') = \sum_{i,j}\mu(A_i\cap A_j')  = \sum_{j=1}^m \mu(  +_{i=1}^n A_i\cap A_j') = \sum_{j=1}^m \mu( A_j').$$

(b.2) Next, to prove that $\bar{\mu}$ is a measure, it suffices to verify nonnegativity, $\bar \mu(\emptyset) = 0$ and the additivity.
The nonnegativity is trivial since $\bar\mu$ being the sum of nonnegative $\mu$ should also be nonnegative. For additivity, we shall prove that the property holds for countably infinite disjoint sets. Given $A = +_{i=1}^\infty A_i$ where $A_i\in \bar{\mathcal S}$, we may first assume $A_i\in \mathcal S$, or we can partition $A_i$ into disjoint sets in $\mathcal S$ but $\sum \mu(A_i)$ remains unchanged.

Then, on the one hand we have 
$$\bar \mu(A)\geqslant \bar\mu(+_{i=1}^n A_i) = \sum_{i=1}^n \bar \mu(A_i) ,\ \forall n\Rightarrow \bar \mu(A)\geqslant \lim_n \sum_{i=1}^n \bar \mu(A_i).$$

On the other hand, since $A\in\mathcal S$, we can write $A = +_{j=1}^m T_j$ where $T_j\in\mathcal S$. Then, $T_j = +_{i=1}^\infty T_j\cap A_i$ together with the condition (ii) in Theorem B implies that $\mu(T_j)\leqslant \sum_{i=1}^\infty \mu(T_j\cap A_i)$. Hence, 
$$\bar \mu(A) = \sum_{j=1}^m \mu(T_j) \leqslant \sum_{j=1}^m\sum_{i=1}^\infty \mu (+_{i=1}^\infty T_j\cap A_i)= \sum_{i=1}^\infty \sum_{j=1}^m \mu (+_{i=1}^\infty T_j\cap A_i)= \sum_{i=1}^\infty \mu(A_i) .$$
Together, we learn $\bar \mu(A)=\sum_{i=1}^\infty \bar \mu(A_i)$ has desired.

## Integration

Let $f$ be a Borel function. We say $\int fd\nu$ exists if at least  $\int f_+(\omega)d\nu$ and $\int f_- (\omega)d\nu$ is finite is finite. And we define 
$$ \int f d\nu = \int f_+(\omega)d\nu -\int f_-(\omega)d\nu .$$

Further, if both $\int f_+(\omega)d\nu$ and $\int f_- (\omega)d\nu$ is finite, we say that $f$ is integrable.

### Fatou's Lemma 

Let $f_1,f_2,\dotsc$ be a sequence of Borel functions on $(\Omega,\mathcal F,\nu)$, and also $f_n\geqslant 0$, then 
$$\int \lim\inf f_n d\nu\leqslant \lim \inf \int f_n d\nu.$$

### Monotone Convergence Theorem

Let $f_1,f_2,\dotsc$ be a sequence of Borel functions on $(\Omega,\mathcal F,\nu)$, and also $0\leqslant f_1\leqslant f_2 \dotsc $ and $\lim f_n = f$ almost everywhere, then 
$$\lim \int f_n d\nu = \int \lim f_n d\nu.$$

### Lebesgue Domianted Convergence Theorem (LDCT)

Let $f_1,f_2,\dotsc$ be a sequence of Borel functions and $\lim f_n = f$ almost everywhere. And if there exists an integrable function $g$ such that $|f_n|\leqslant g$, then 
$$\lim \int f_n d\nu = \int \lim f_n d\nu.$$

### Change of Variables

Let $f$ be measurable from $(\Omega, \mathcal F,\nu)$ to $(\Lambda, \mathcal G)$ and $g$ be Borel on $(\Lambda, \mathcal G)$. Then, 
$$\int_\Omega g\circ fd\nu = \int_\Lambda g d(\nu \circ f^{-1}).$$

### Expectance

Let $X$ be a random varaiable (Borel function) on $(\Omega,\mathcal F,\mathbb P)$, then we define the expectance of $X$ by 
$$\mathbb E(X) = \int_{\Omega}Xd\mathbb P  = \int_{\mathbb R}xd\mathbb P\circ X^{-1} = \int_{\mathbb R}xd\mathbb P_X$$

## Derivative

### Absolute Continuity 

Let $\nu$ and $\lambda$ be two measures on $(\Omega,\mathcal F)$ and if they have the following property:
$$\nu(A)=0\Rightarrow \lambda(A)=0\quad \forall A\in \mathcal F,$$
then we write $\lambda \ll \nu$ and say that $\lambda$ is absolutely continuous with respect to $\nu$.


**Theorem** Let $f$ be a nonnegative Borel function on $(\Omega,\mathcal F,\nu)$. Denote $\lambda(A)=\int_A fd\nu$ for $A\in\mathcal F$. Then $\lambda\ll \nu$.

One would then question whether an absolutely continuous function must be in the form of integral. The answer is yes and we have the following theorem.



### Radon-Nikodym Theorem 

Let $\nu$ and $\lambda$ be two measures on $(\Omega,\mathcal F)$ and $\nu$ be $\sigma$-finite. If $\lambda \ll \nu$, then there exists a unique nonnegative Borel function $f$ on $\Omega$ such that $\lambda (A) = \int_A fd\nu$. And we call $f$ the Radon-Nikodym derivative / density of $\lambda$ with respect to $\nu$. We write $f = \dfrac{d\lambda}{d\nu}$.
