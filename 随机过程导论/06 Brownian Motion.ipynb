# 6 Brownian Motion

If $B = \{B_t:\ t\geqslant 0\}$ is a real-valued stochastic process with the following properties:

1. $B_0 = 0$.
2. Independent increments: for any $0\leqslant t_1 < t_2 < \cdots < t_n$, the random variables $B_{t_1}, B_{t_2} - B_{t_1}, \ldots, B_{t_n} - B_{t_{n-1}}$ are independent.
3. Stationary and normal increments: for any $0\leqslant s < t$, the random variable $B_t - B_s$ is normally distributed with mean $0$ and variance $t-s$, i.e. $B_t - B_s \sim N(0, t-s)$.

then $B$ is called a *Brownian motion*, also known as a *Wiener process*.

In particular, $B_t \sim N(0,t)$ is Gaussian.

## Statistics

### Covariance

**Theorem** For $t,s\geqslant 0$, the covariance is given by ${\rm Cov}(B_t B_s) = \min(t,s)$. And its correlation is given by $\rho_{t,s} = \frac{\min(t,s)}{\sqrt{ts}}$.

**Proof** Without loss of generality, assume $s\leqslant t$. Then

$${\rm Cov}(B_t B_s) =\mathbb E(B_tB_s)=\mathbb E((B_t - B_s)B_s) + \mathbb E(B_s^2) = \mathbb E(B_s^2) = s.$$

The correlation is obtained by dividing the covariance by the $\sqrt{{\rm Var}(B_t){\rm Var}(B_s)} = \sqrt{ts}$.


### Joint Distribution

**Theorem** For $t_1,\dotsc,t_n$, the joint distribution of $(B_{t_1},\dotsc,B_{t_n})$ is a multivariate normal distribution with mean $\mathbf 0$ and covariance matrix $\Sigma$ where $\Sigma_{i,j} = \min(t_i,t_j)$.

<br>

For a stochastic process, if every finite subset of the process is jointly normal, then the process is called a *Gaussian process* (GP). This theorem shows that Brownian motion is a Gaussian process. The following theorem shows when a Gaussian process is a Brownian motion.


**Theorem** If $B$ is a Gaussian process and $\mathbb E(B_t) = 0$ and ${\rm Cov}(B_t B_s) = \min(t,s)$, then $B$ is a Brownian motion.

**Proof** We need to show that $B$ satisfies the three properties of a Brownian motion. $B_0\sim N(0,0)$ so $B_0 = 0$. For the second property, let $0\leqslant t_1 < t_2 < \cdots < t_n$. Then we see that $B_{t_1}, B_{t_2} - B_{t_1}, \ldots, B_{t_n} - B_{t_{n-1}}$ are jointly normal. For $i<j$, the covariance is given by

$${\rm Cov}\left(B_{t_j} - B_{t_{j-1}}, B_{t_i} - B_{t_{i-1}}\right)=t_i - t_{i-1}-t_i + t_{i-1}=0.$$

Thus, the random variables are independent. Lastly, the third property is trivial.


### Transformation

If $B$ is a Brownian motion, then the following three new processes are also Brownian motions:

1. Fix $t_0$, and $x_t = B_{t_0 + t} - B_{t_0}$ is a Brownian motion.
2. Fix $c$, and $x_t = \frac{1}{\sqrt{c}}B_{c t}$ is a Brownian motion.
3. $x_t = \left\{\begin{array}{ll}0 & t= 0\\ tB_{\frac 1t} & t>0\end{array}\right.$ is a Brownian motion.

**Proof** The first and the second are trivial. We only prove the third. First, we show that $x_t$ is a Gaussian process, which suffices to check that the joint distribution of any finite subset is jointly normal. Using Cramer-Wold device, it is equivalent to show that $\sum a_tx_t$ is normal for any $a_1,\dotsc,a_n$. This surely holds.

Next, for $s\leqslant t$, $\mathbb E(x_t) = \mathbb E(tB_{\frac 1t}) = 0$. And

$${\rm Cov}(x_t x_s) = \mathbb E\left(tB_{\frac 1t}sB_{\frac 1s}\right) = ts\mathbb E\left(B_{\frac 1t}B_{\frac 1s}\right) =ts \min\left(t^{-1},s^{-1}\right) = \min(s,t).$$

### Conditional Distribution

* Prediction: For $0\leqslant s < t$, the conditional distribution of $B_t$ given $B_s$ is $f_{B_t|B_s}(y|x) \sim N(x, t-s)$.
* Backcasting: For $0\leqslant s < t$, the conditional distribution of $B_s$ given $B_t$ is $f_{B_s|B_t}(x|y) \sim N\left(\frac{s}{t}y, \frac{s(t-s)}{t}\right)$.
* Between: For $t_1<t<t_2$, the conditional distribution of $B_t$ given $B_{t_1},B_{t_2}$ is $f_{B_t|B_{t_1},B_{t_2}}(x|x_1,x_2)\sim N\left(x_1+\frac{(x_2-x_1)(t-t_1)}{t_2-t_1},\frac{(t_2-t)(t-t_1)}{t_2-t_1}\right)$.
  
**Proof** The first is trivial. For the second, fixing $t$ and $y$ we have

$$f_{B_s|B_t}(x|y)\propto {f_{B_s,B_t}(x,y)} = \frac{1}{\sqrt{2\pi s}}\exp\left(-\frac{x^2}{2s}\right)\frac{1}{\sqrt{2\pi (t-s)}}\exp\left(-\frac{(y-x)^2}{2(t-s)}\right).$$


It should be a normal distribution. The variance is the half the denominator of $x^2$ term, which is $\frac{1}{\frac{1}{s}+\frac{1}{t-s}} = \frac{s(t-s)}{t}$. The mean corresponds to the coefficient of $x$ multiplies the variance, which is $\frac{y}{t-s}\cdot\frac{s(t-s)}{t}=\frac sty$.

The third and the second are actually equivalent if we take the transformation $B(t) = \tilde B(t-t_1) + B(t_1)$.

## Simple Random Walk

We can use simple random walk to approximate Brownian motion. Let $X_1,X_2,\dotsc$ be i.i.d. random variables with $\mathbb P(X_i = 1) = \mathbb P(X_i = -1) = \frac 12$. Let $S_n = \sum_{i=1}^n X_i$. Then $S_n$ is a simple random walk.

Now we consider the following sequence of random variable: $\tilde B_n(t) = \frac{S_{nt}}{\sqrt n}$ for $t\in \mathbb N$. Then ${\rm Var}(\tilde B_n(t)) = \frac 1n \cdot nt = t$. Hence by Linderberg-Feller's central limit theorem,

$$S_n = \sum_{i=1}^n Z_i \xrightarrow{d} N(0,t).$$ 

<br>

Therefore we have the following relation: if $S_n=\sum X_n$ is a simple symmetric random walk with step size $1$, then $\tilde B(t)=\frac{S_{[nt]}}{\sqrt n}$ converges to a Brownian motion when $n\to\infty$.



```python
from matplotlib import pyplot as plt
import numpy as np

def gen_brownian(t = 1., num = 1000, n = 1000, seed = 0):
    if seed is not None:
        np.random.seed(seed)
    x = np.where(np.random.random(round(n*t)+1) < .5, 1, -1)
    s = np.cumsum(x)
    ts = np.linspace(0, t, num)
    Bt = s[np.floor(n * ts).astype('int')] / np.sqrt(n)
    return ts, Bt

ts, Bt = gen_brownian(n = 10000)
plt.plot(ts, Bt)
plt.show()
```

## Variations

### Reflected Brownian Motion

Define $x_t = |B_t|$. Then $x_t$ is called a *reflected Brownian motion*. Easy to see that

$$\begin{aligned}
\mathbb E(x_t) &=2 \int_0^\infty x\frac{1}{\sqrt{2\pi t}}\exp\left(-\frac{x^2}{2t}\right)dx = \sqrt{\frac{2t}{\pi}}
\\ {\rm Var}(x_t) &= \mathbb E(x_t^2) - \mathbb E(x_t)^2 = t-\frac{2t}{\pi} = \frac{\pi - 2}{\pi }t.
\end{aligned}$$


### Geometric Brownian Motion

Define $x_t = e^{\alpha B_t}$. Then $x_t$ is called a *geometric Brownian motion*. Easy to see that

$$\begin{aligned}
\mathbb E(x_t) &= e^{\alpha^2t/2}.
\\ {\rm Var}(x_t) &= e^{2\alpha^2t}- e^{\alpha^2t}
\end{aligned}$$

### Brownian Bridge

Define $B_t^0 = B_t - tB_1$. Then $B_t^0$ is called a *Brownian bridge*. Clearly that $B_t^0\sim B_t|(B_{1}=0) \sim N(0,t(1-t))$. The covariance is given by

$$\begin{aligned}
{\rm Cov}(B_t^0 B_s^0) &= \mathbb E(B_t^0 B_s^0) = \mathbb E(B_tB_s - tB_s - sB_t + ts) = \mathbb E(B_tB_s) - t\mathbb E(B_s) - s\mathbb E(B_t) + ts
\\ &= \min(t,s) - ts.
\end{aligned}$$

### Integrated Brownian Motion

Define $x_t = \int_0^t B_s ds$. Then $x_t$ is called an *integrated Brownian motion*. Then $x_t$ is also Gaussian. And for $s\leqslant t$,

$$\begin{aligned}
\mathbb E(x_t) &= \int_0^t \mathbb E(B_s) ds = 0\\ 
{\rm Cov}(x_tx_s) &= \int_0^t\int_0^s \min \{u,v\}dudv = \int_0^s \int_0^u vdvdu + \int_0^s \int_u^t udvdu
\\ &= \int_0^s \frac 12u^2du +\int_0^s u(t-u)du = \frac12 {ts^2}-\frac 16 {s^3}\quad\quad (s\leqslant t)\\
{\rm Var}(x_t) &= \frac 13 t^3.
\end{aligned}$$


### Running Maximum

Define $M_t = \sup_{0\leqslant s\leqslant t} B_s$. Then $M_t$ is called the *maximum process*.

**Theorem** $M_t$ has the same distribution as the reflected Brownian motion, $M_t\sim |B_t|$.

**Proof** Since $M_t\geqslant B_0 = 0$, it is always non-negative. And for $x\geqslant 0$,

$$\mathbb P(M_t>x)=\mathbb P(M_t>x, B_t>x) +\mathbb P(M_t>x, B_t\leqslant x)=\mathbb P(B_t>x) +\mathbb P(M_t>x, B_t\leqslant x).$$

For sequences that $M_t>x$ but $B_t\leqslant x$, we consider the following reflection mapping: Let $\tau =\sup \{s:\ s\leqslant t,\ B_s\geqslant x\}$. We define $\tilde B_s = \left\{\begin{array}{ll}B_s & s \leqslant\tau \\ 2x - B_s & \tau  < s\leqslant t\end{array}\right.$. Then for the new sequence $\tilde B_t$, we have $\tilde B_t\geqslant x$. Also, it is easy to verify $\tau $ has a inverse, so it is bijective.

Now we have found a bijection $B\mapsto\tilde B$, a mapping between $\{M_t>x,B_t\leqslant x\}$ and $\{B_t\geqslant x\}$. Therefore, $\mathbb P(M_t>x)$ is twice $\mathbb P(B_t\geqslant x)$, the outskirt of a Gaussian. Hence $M_t\sim |B_t|$.


```python
# example to illustrate the reflection mapping
ts, Bt = gen_brownian(t = 1, n = 10000, seed = 8)

x_split = 1.09
tau = np.where(Bt >= x_split, ts, 0).max()
print('(Tau, x) = (%.2f, %.2f)'%(tau, x_split))

reflected_Bt = np.where(ts <= tau, Bt, 2*x_split - Bt)

plt.figure(figsize = (10, 6))
plt.plot(ts, Bt, label = 'Bt', alpha = .7)
plt.plot(ts, reflected_Bt, label = 'Reflected Bt', alpha = .7)
plt.axhline(x_split, color = 'k', linestyle = '--', alpha = .4)
plt.axvline(tau, color = 'k', linestyle = '--', alpha = .4)
plt.scatter(tau, x_split, color = 'r', s = 50, label = '($\\tau$, $x$) = (%.2f, %.2f)'%(tau, x_split), zorder = 3)
[plt.title('Reflection Mapping'), plt.legend()]
plt.show()
```

### Joint Distribution of Brownian Motion and Its Maximum

Let $B_t$ be the Brownian motion and $M_t = \sup_{0\leqslant s\leqslant t}B_s$ be its maximum process. Then the joint distribution of $(B_t,M_t)$ is given by $(y\geqslant \max\{0,x\}$)

$$\mathbb P(B_t\leqslant x, M_t\leqslant y) = \Phi\left(\frac{x}{\sqrt t}\right)+\Phi\left(\frac{2y-x}{\sqrt t}\right)-1$$

with P.D.F. given by

$$f_{B_t,M_t}(x,y) = \sqrt{\frac{2}{\pi t^3}}(2y-x)\exp\left\{-\frac{(2y-x)^2}{2t}\right\}.$$


**Proof** 

For sequences that $M_t>y\geqslant x$ but $B_t\leqslant x$, we consider the following reflection mapping: Let $\tau =\sup \{s:\ s\leqslant t,\ B_s\geqslant x\}$. We define $\tilde B_s = \left\{\begin{array}{ll}B_s & s \leqslant\tau \\ 2y - B_s & \tau  < s\leqslant t\end{array}\right.$. Then for the new sequence $\tilde B_{\cdot}$, we have $\tilde B_t\geqslant 2y- x$. Also, it is easy to verify $\tau $ has a inverse, so it is bijective.

Now we have found a bijection $B\mapsto\tilde B$, a mapping between $\{M_t>y,B_t\leqslant x\}$ and $\{B_t\geqslant 2y-x\}$. Therefore, 

$$\mathbb P(B_t\leqslant x, M_t>y) =\mathbb P(B_t\geqslant 2y-x) = 1-\Phi\left(\frac{2y-x}{\sqrt t}\right).$$

Whilst $\mathbb P(B_t\leqslant x) = \Phi\left(\frac{x}{\sqrt t}\right)$, so

$$\mathbb P(B_t\leqslant x, M_t\leqslant y) = \Phi\left(\frac{x}{\sqrt t}\right)+\Phi\left(\frac{2y-x}{\sqrt t}\right)-1.$$

Its density is given by $\frac{\partial^2 P}{\partial x\partial y} = \frac{\partial^2}{\partial x\partial y}\Phi\left(\frac{2y-x}{\sqrt t}\right)= \sqrt{\frac{2}{\pi t^3}}(2y-x)\exp\left\{-\frac{(2y-x)^2}{2t}\right\}$.


```python
import sympy as sp
x, y, t, u = sp.symbols('x y t u')
Phi = sp.Integral(sp.exp(-u**2 / (2)) / sp.sqrt(2*sp.pi), (u, -sp.oo, x))
pdf = Phi.subs(x, (2*y-x) / sp.sqrt(t)).diff(x).diff(y).together()
print(pdf)
```

### Hitting Time

Let $a>0$ to be a constant. Define $T_a = \inf\{t\geqslant 0: B_t = a\}$. Then $T_a$ is called the *hitting time* of $a$.

**Theorem** $T_a\leqslant t$ if and only if $M_t\geqslant a$. Hence $T_a$ follows a Levy distribution: $f_{T_a}(t) = \frac{a}{\sqrt{2\pi t^3}}e^{-\frac{a^2}{2t}}\ (t>0)$.

Also, the mean of the distribution is infinite.

**Proof** The first statement is trivial. Now we derive its P.D.F..

$$\mathbb P(T_a\leqslant t) = \mathbb P(M_t\geqslant a) = \mathbb P(|N(0,t)|\geqslant a) =  2\int_a^\infty \frac{1}{\sqrt{2\pi t}}e^{-\frac{u^2}{2t}}du=\sqrt{\frac{2}{\pi}}\int_{\frac{a}{\sqrt t}}^\infty  e^{-\frac{v^2}{2}}dv.$$

Taking derivative,

$$f_{T_a}(t) =  -\sqrt{\frac{2}{\pi}}e^{-\frac{a^2}{2t}}\cdot \left(-\frac{a}{2t^\frac32}\right)= \frac{a}{\sqrt{2\pi t^3}}e^{-\frac{a^2}{2t}}.$$

As for the mean,

$$\mathbb E(T_a) = \int_0^\infty  \frac{a}{\sqrt{2\pi t}}e^{-\frac{a^2}{2t}}dt\geqslant \int_{1}^\infty\frac{a}{\sqrt{2\pi t}}e^{-\frac{a^2}{2}}dt=\infty.$$


### Absorbed Brownian Motion

Let $a>0$ be a constant, and assume the Brownian motion is absorbed when reaching $a$. That is, 

$$Z(t) = \left\{\begin{array}{ll}B_t & B_t\leqslant a \\ a & B_t > a\end{array}\right..$$

Then, for $x<a$, $\mathbb P(Z(t)\leqslant x) = \mathbb P(B_t\leqslant x, \ M_t\leqslant a)$ is a joint distribution of a Brownian motion and its maximum. For $x = a$, $\mathbb P(Z(t) =a) = \mathbb P(M_t\geqslant a) = 2(1 - \Phi(\frac{x}{\sqrt t}))$.
