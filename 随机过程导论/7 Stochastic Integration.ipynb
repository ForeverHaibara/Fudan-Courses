## Riemann-Stieltjes Integral

The Riemann-Stieltjes integral $\displaystyle \int_a^b f(x) dG(x)$ is defined as follows: for any partition $P = \{a = x_0 < x_1 < \cdots < x_n = b\}$ of $[a,b]$, and any choice of points $\{t_i\}_{i=1}^n$ in $[x_{i-1}, x_i]$, we define the Riemann-Stieltjes sum as

$$s_n = \sum_{i=1}^n f(t_i) \left(G(x_i) - G(x_{i-1})\right)$$

If $s_n\rightarrow s$ when $n\rightarrow\infty$ for any partition $x_i$ such that $\max|x_{i+1}-x_i|\rightarrow 0$ and for any choice of $t_i$, then we say that $\displaystyle \int_a^b f(x) dG(x) = s$.

### Bounded Variation

Let $G$ be a function on $[a,b]$. We say that $G$ is of bounded variation on $[a,b]$ if there exists a constant $M$ such that for any partition $P = \{a = x_0 < x_1 < \cdots < x_n = b\}$ of $[a,b]$, we have

$$\sum_{i=1}^n \left|G(x_i) - G(x_{i-1})\right| \leqslant M$$

**Theorem** If $G$ is of bounded variation on $[a,b]$, then the Riemann-Stieltjes integral exists for any continuous function $f$ on $[a,b]$.

### Brownian Motion

**Theorem** Brownian motion $B(t)$ is not of bounded variation on any interval $[a,b]$ almost surely.

**Proof** Without loss of generality, we assume $a=0$ and $b=1$. We slice $[0,1]$ into $n$ equal parts, and consider the sum

$$S_n = \sum_{i=1}^n \left|B\left(\frac{i}{n}\right) - B\left(\frac{i-1}{n}\right)\right|$$

Recall $B(\frac in) - B(\frac{i-1}{n}) \sim N(0, \frac{1}{n})$, so

$$\mathbb E(S_n) = n\cdot \sqrt{\frac{2}{\pi n}} \rightarrow \infty\quad\quad{\rm and}\quad\quad 
{\rm Var}(S_n) = n\cdot \frac{\pi - 2}{\pi n} = \frac{\pi - 2}{\pi}.$$


For arbitrary $M$, we have

$$\mathbb P(S_n\leqslant M) \leqslant \mathbb P\left(\left|S_n - \mathbb E(S_n)\right| \leqslant |M - \mathbb E(S_n)|\right) \leqslant \frac{{\rm Var}(S_n)}{\left(M - \mathbb E(S_n)\right)^2} \rightarrow 0.$$

### Integration by Parts


Although the Brownian motion $B(t)$ is not of bounded variation, we can still define the Riemann-Stieltjes integral $\displaystyle \int_0^t f(s) dB(s)$ for any bounder-variation function $f$ by integrating by parts. 

$$\int_0^t f(s) dB(s) = f(t)B(t) - \int_0^t B(s) df(s).$$

Note that $B(t)$ is continuous almost surely, so the latter integral is well-defined as long as $f$ is of bounded-variation. The details are demonstrated in the next section.

## Itô Integral

We can integrate random processes with respect to Brownian motion. Suppose $B(t)$ is a Brownian motion, and $f(t)$ is a random process independent with $\{B(s)\}_{s> t}$ and $\int_0^\infty \mathbb E(f(s))^2ds<\infty$. Now we partition $[0,t]$ into $P_n = \{0 = s_0 < s_1 < \cdots < s_n = t\}$, and define the Itô sum by

$$S_n = \sum_{i=1}^n f(s_{i-1})\left(B(s_i) - B(s_{i-1})\right)$$

If there exists random variable $x_t$ such that $\mathbb E|s_n - x_t|^2\rightarrow 0$ for any partition such that  $\max|s_{i+1}-s_i|\rightarrow 0$ when $n\rightarrow\infty$, then we say that $\displaystyle \int_0^t f(s) dB(s) = x_t$. This is the Itô integral (伊藤积分).

### Properties

Fixing $t$, assume the Itô integral $X_t = \displaystyle \int_0^t f(s) dB(s)$ exists, then

* Expectation: $\mathbb E\left(X_t\right) = 0$.
* Variance (**Itô isometry**): $\mathbb E\left(X_t^2\right) =\displaystyle  \int_0^t \mathbb E(f(s))^2 ds$.
* Linearity: $\displaystyle \int_0^t \left(a f(s) + b g(s)\right) dB(s) = a\int_0^t f(s) dB(s) + b\int_0^t g(s) dB(s)$.
* Additivity: $\displaystyle \int_0^t f(s) dB(s) + \int_t^T f(s) dB(s) = \int_0^T f(s) dB(s)$.

**Proof** We will only derive the expectation and variance. As for expectation, note that by definition we require $f(s_{i-1})$ to be independent with $B(s)\ (s> s_{i-1})$, and we thus have

$$\mathbb E\left\{ f(s_{i-1})\left(B(s_i) - B(s_{i-1})\right)\right\} = 0\ \Rightarrow\ \mathbb E S_n = 0.$$

By Lesbegue's dominated convergence theorem, we have that

$$ \mathbb E\displaystyle \int_0^t f(s) dB(s) = \mathbb E\lim S_n = \lim \mathbb E S_n = 0.$$

<br>

For the variance, using

$$\mathbb E\left\{f(s_{i-1})\left(B(s_i) - B(s_{i-1})f(s_{j-1})(B(s_j) - B(s_{j-1}))\right) \right\}= \left\{\begin{array}{ll} \mathbb E f(s_{i-1})^2 \cdot (s_i - s_{i-1}) & i = j \\ 0 & i\neq j\end{array}\right.$$

we yield

$$\mathbb E(X_t^2)= \mathbb E \lim S_n^2 =\lim \mathbb ES_n^2 = \lim \sum_{i=1}^n \mathbb E f(s_{i-1})^2 \cdot (s_i - s_{i-1}) = \int_0^t \mathbb E f(s)^2 ds.$$

### Examples

1. $\displaystyle \int_0^t B_sdB_s = \frac12 (B_t^2 - t)$
2. $\displaystyle \int_0^t s dB_s = tB_t - \displaystyle \int_0^t B_sds$.

**Proof** For the first, we have

$$\begin{aligned}2\sum B(s_{i-1})\left(B(s_i) - B(s_{i-1})\right) &=  \sum \left(B(s_i)^2  -B(s_{i-1})^2\right) - \sum \left(B(s_i) - B(s_{i-1})\right) ^2\\ & = B_t^2 - \sum \left(B(s_i) - B(s_{i-1})\right) ^2\end{aligned} $$

Now we note that $\left(B(s_i) - B(s_{i-1})\right) ^2$ are independent $(s_i - s_{i-1})\chi^2_1$ random variables and the $L_2$ distance is bounded by

$$\begin{aligned}\mathbb E\left\{ \left[\sum (s_i - s_{i-1})\chi^2_1 - t\right]^2\right\} & =\sum \text{Var}\left( (s_i - s_{i-1})\chi^2_1\right)=\text{Var}\chi^2_1\sum (s_i - s_{i-1})^2
\\ &\leqslant \text{Var}\chi^2_1\cdot \sup_{i}\left\{s_i - s_{i-1}\right\}\cdot t \rightarrow 0.
\end{aligned}$$

Therefore, the integral converges to $\frac12 (B_t^2 - t)$ in $L_2$ sense.

<br>

For the second, using Abel summation formula, we have

$$\sum_{i=1}^n s_{i-1}\left(B(s_i) - B(s_{i-1})\right) = -\sum_{i=1}^{n-1} (s_i - s_{i-1})B(s_{i}) + s_nB(s_n) - s_0B(s_0) = tB_t -\sum_{i=1}^{n-1} (s_i - s_{i-1})B(s_{i}).$$

### Itô Formula

Suppose $B_t$ is a standard Brownian motion, and $f(t,x)$ is twice continuously differentiable function with respect to $x$ and continuously differentiable with respect to $t$. Then we have

$$\text{d}f(t, B_t) = \left(f_t'(t, B_t) + \frac12 f_{xx}''(t, B_t)\right)\text{d}t + f_x'(t, B_t)\text{d}B_t.$$

### Example

1. $\displaystyle \int_0^t B_s^2 dB_s  = \frac13 B_t^3 - \frac13 \int_0^t B_sds$

**Proof** Applying Itô formula on $f(t,x) = \frac13x^3$, we have $\text{d}\frac{B_t^3}{3} = B_t\text{d}t +  B_t^2\text{d}B_t$. Thus,

$$  \int_0^t B_s^2 dB_s =\int \text{d}\frac{B_s^3}{3} - \int_0^t B_sds = \frac13 B_t^3 - \frac13 \int_0^t B_sds.$$
