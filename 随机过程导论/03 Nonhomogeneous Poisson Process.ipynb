# 03 Nonhomogeneous Poisson Process

## Nonhomogeneous Poisson Process

Nonhomogeneous Poisson process (NHPP) has variant rate parameter $\lambda$. Its "rare event" property is redefined by

$$\begin{aligned}\mathbb P\left(N(t+\Delta t) - N(t)=1\right) &= \lambda(t)\cdot\Delta t + o(\Delta t)
\\ \mathbb P\left(N(t+\Delta t) - N(t)\geqslant 2\right) &=  o(\Delta t)
\end{aligned}$$

Here $\lambda(t)$ is a function of $t$.

It therefore no longer preserves stationary increments. But we still require independent increments. 

**Theorem** If we define $m(s,t)=\int_s^t \lambda (u)du$, we have

$$\mathbb P(N(s,t]=k)  = \frac{m(s,t)^{k}}{k!}e^{-m(s,t)}.$$

**Proof** Denote $P_k(t) = P(N(t)=k)$. Note that $P_0(0)=1$. Then we have
$$\begin{aligned}P_0(t+\Delta t) &= P(N(t+\Delta t) = 0)=\mathbb P(N(t)=0,\ N(t+\Delta t)-N(t)=0)
\\ &= \mathbb P(N(t)=0)\cdot \mathbb P(N(t,t+\Delta t]=0)
\\ &= P_0(t)\cdot (1 -\lambda (t)\Delta t+o(\Delta t)).
\end{aligned}$$

Thus we see that $\mathbb P_0'(t) = -\lambda(t)P_0(t)$ by the definition of derivative, which has solution $P_0(t) = e^{-m(0,t)}$. Similarly, we have
$$\begin{aligned}P_k(t+\Delta t)& = \mathbb P(N(t)=k)\cdot \mathbb P(N(t,t+\Delta t]=0)+\mathbb P(N(t)=k-1)\cdot \mathbb P(N(t,t+\Delta t]=1)+o(\Delta t)
\\ &= P_k(t)+\left(-P_k(t)+ P_{k-1}(t)\right)\lambda (t)\cdot\Delta t + o(\Delta t).
\end{aligned}$$

Then we can prove by induction that $P_k(t) = \frac{m(0,t)^{k}}{k!}e^{-m(0,t)}$. Finally, the proof can be generalized to arbitrary initial start time $s$ so $m(0,t)$ is replaced by $m(s,t)$ for $P(N(s,t]=k)$.

## Arrival and Waiting

### Arrival Time

If $N(t)=k$, then there are $k$ events before time $t$. Denote $s_i$ be the occurrence time of the $i$-th event. We call it the arrival time.

**Theorem** If $N(t)$ flows a nonhomogeneous Poisson process with rate $\lambda$, then $s_n$ has P.D.F. $f_{s_n}(t) = \frac{\lambda(t)m(0,t)^{n-1}}{(n-1)!}e^{-m(0,t)}$.

**Proof** We have
$$\begin{aligned}\mathbb P(s_n\leqslant t) &= \mathbb P(N(t)\geqslant n) = 1-\sum_{k=0}^{n-1} \frac{m(0,t)^k}{k!}e^{-m(0,t)}.\end{aligned}$$

Then we can take the derivative to get the P.D.F. of $s_n$ that
$$\begin{aligned}f_{s_n}(t) &= \frac{d}{dt}\sum_{k=n}^{\infty} \frac{m(0,t)^k}{k!}e^{-m(0,t)}=
\sum_{k=n}^{\infty}\frac{m(0,t)^k}{k!}\left(\frac{k}{m(0,t)}-1\right)\lambda (t)e^{-m(0,t)}=\frac{\lambda(t)m(0,t)^{n-1}}{(n-1)!}e^{-m(0,t)}.\end{aligned}$$

### Waiting Time

The waiting time is no longer independent with each other in a nonhomogeneous Poisson process. But we can still calculate the P.D.F. of the waiting time.


**Theorem** If $N(t)$ flows a nonhomogeneous Poisson process with rate $\lambda$, then for $n\geqslant 2$, the P.D.F. of the waiting time $x_n=s_n-s_{n-1}$ is given by 

$$f_{x_n}(t)=\frac{1}{(n-2)!}\int_0^\infty \lambda(s)m(0,s)^{n-2}e^{-m(0,s+t)} ds.$$

**Proof** For $n\geqslant 2$,

$$\begin{aligned}\mathbb P(x_n>t) &= \mathbb P(s_n-s_{n-1}>t) = \int_0^\infty \mathbb P(s_n>t+s|s_{n-1}=s)f_{s_{n-1}}(s)ds\\ &=\int_0^\infty \mathbb P(N(t+s) - N(s) = 0)f_{s_{n-1}}(s)ds\\ &=  \int_0^\infty {e^{-m(s,s+t)}}{ }\cdot\frac{\lambda(s)m(0,s)^{n-2}}{(n-2)!}e^{-m(0,s)} ds.\end{aligned}$$

## Time Translation

In the following, we consider a nonhomogeneous Poisson process with rate $\lambda(t)$, denoted by $\tilde N(t)$. We define $m(t) = \int_0^\infty \lambda(s)ds$ and its inverse function to be $m^{-1}(\cdot)$.

**Theorem** Let $N(t) = \tilde N(m^{-1}(t))$, then $N(t)$ is a homogeneous Poisson process with rate $1$.

**Proof** 

Firstly, $N(0)= \tilde N(0)=0$. 

Secondly, the "rare event" property is preserved by
$$\mathbb P(N(t+\Delta t) -N(t)=k) = \mathbb P\left(\tilde N(m^{-1}(t+\Delta t)) - \tilde N(m^{-1}(t))=k\right).$$

Note that $m'(t)=\lambda (t)$, so we have $\frac{\partial }{\partial t}m^{-1}(t) = \frac{1}{\lambda(m^{-1}(t))}$. Thus,

$$\tilde\Delta t = m^{-1}(t+\Delta t)-m^{-1}(t) = \frac{1}{\lambda (m^{-1}(t))}\Delta t+ o(\Delta t).$$

And we have
$$\begin{aligned}\mathbb P\left(\tilde N(m^{-1}(t+\Delta t)) - \tilde N(m^{-1}(t))=1\right) &= \lambda(m^{-1}(t))\cdot \tilde \Delta t+o(\tilde \Delta t)=\Delta t+ o(\Delta t)
\\  \mathbb P\left(\tilde N(m^{-1}(t+\Delta t)) - \tilde N(m^{-1}(t))\geqslant 2\right) &= o(\tilde \Delta t)= o(\Delta t).\end{aligned}$$

Thirdly, the stationarity is shown by $(s<t)$
$$\begin{aligned}\mathbb P(N(t)-N(s)=k) &= \mathbb P\left(\tilde N(m^{-1}(t)) - \tilde N(m^{-1}(s)) = k\right)
\\ &= \frac{m(m^{-1}(s),m^{-1}(t))^{k}}{k!}e^{-m(m^{-1}(s),m^{-1}(t))}=\frac{(t-s)^k}{k!}e^{-(t-s)}.
\end{aligned}$$

## Compound Poisson Process

Assume $N(t)$ is a homogeneous Poisson process with rate $\lambda$. Suppose $\xi_1,\xi_2,\dotsc$ are i.i.d. random variables and they are all independent with $N(t)$. Define $Z(t) = \sum_{i=1}^{N(t)}\xi_i$. Then $Z(t)$ is called a compound Poisson process.

### Examples

For example, the number of customers arriving in a shop can be simulated by a Poisson process. And the total amount of money they spend can be simulated by a compound Poisson process.

As for another example, let $\xi_i\sim B(p)$ is Bernoulli distributed. Then $Z(t)$ is the number of times that a Bernoulli trial is successful, which is a decomposition of the Poisson process.


### Properties

Aside from $Z(0)=0$, compound Poisson process $Z(t)$ has independent and stationary increments.

#### Independent Increments

**Theorem** for $s<t$, $Z(t)-Z(s)$ is independent with $Z(s)$.

**Proof** Using the method of characteristic function, we have

$$\begin{aligned}\phi_{Z(t)-Z(s),Z(s)}(a,b) &= \mathbb E\exp\left\{ia(Z(t)-Z(s))+ibZ(s)\right\}
\\ &= \mathbb E_{N(t),N(s)}\mathbb E_{\xi}\exp\left\{ia(Z(t)-Z(s))+ibZ(s)\right\}
\\ &= \mathbb E_{N(t),N(s)}\mathbb E_{\xi}\exp\left\{ia\sum_{k>N(s)}^{N(t)}\xi_k+ib\sum_{k=1}^{N(s)}\xi_k\right\}\\ &= \mathbb E_{N(t),N(s)}\left\{\phi_\xi(a)^{N(t)-N(s)}\phi_\xi(b)^{N(s)}\right\}\\ &= \mathbb E_{N(t-s),N(s)}\left\{\phi_\xi(a)^{N(t-s)}\phi_\xi(b)^{N(s)}\right\}
\\ &= \mathbb E_{N(t-s)}\left\{\phi_\xi(a)^{N(t-s)}\right\}\mathbb E_{N(s)}\left\{\phi_\xi(b)^{N(s)}\right\}.
\end{aligned} $$

It is in the form of $f_1(a)f_2(b)$, which means $Z(t)-Z(s)$ is independent with $Z(s)$.

#### Stationary Increments

**Theorem** For $s<t$, $Z(t)-Z(s)$ has the same distribution as $Z(t-s)$.

**Proof** Using the method of characteristic function combined with the details above,

$$\phi_{Z(t)-Z(s)}(a) = \mathbb E\exp\left\{ia(Z(t)-Z(s))\right\} = \mathbb E_{N(t-s)}\left\{\phi_\xi(a)^{N(t-s)}\right\}.$$

It only depends on $t-s$, so $Z(t)-Z(s)$ has the same distribution as $Z(t-s)-Z(0)=Z(t-s)$.


### Statistics

Now we assume the original Poisson process $N(t)$ has rate $\lambda$ and $\xi_i$ has mean $\mu$ and variance $\sigma^2$. 

**Theorem** $\mathbb E(Z(t)) = \lambda t\mu$ and ${\rm Var}(Z(t))=\lambda t(\mu^2 +\sigma^2)$.

**Proof** The mean is given by

$$\mathbb E(Z(t)) = \mathbb E_{N(t)}\mathbb E_\xi \sum_{i=1}^{N(t)}\xi_i = \mathbb E_{N(t)}N(t)\mu = \lambda t\mu.$$
 
To compute the variance, we first compute the second moment

$$\begin{aligned}\mathbb E(Z^2(t)) &= \mathbb E_{N(t)}\mathbb E_\xi \left(\sum_{i=1}^{N(t)}\xi_i\right)^2
=\mathbb E_{N(t)}\left\{N(t)(\mu^2+\sigma^2)+N(t)(N(t)-1)\mu^2\right\} 
\\ &= \lambda t(\mu^2+\sigma^2)+(\lambda^2t^2+\lambda t-\lambda t)\mu^2=\lambda t(\mu^2+\sigma^2)+
\lambda^2t^2\mu^2.
\end{aligned}$$

Therefore the variance is the difference between the second moment and the square of the mean, which is $\lambda t(\mu^2+\sigma^2)$.

<br>

NOTE: Since the Poisson process requires the mean and the variance are equal, we can see that the compound Poisson process is generally NOT a Poisson process.


#### Characteristic Function

**Theorem** The characteristic function of $Z(t)$ is given by $\phi_{Z(t)}(a)=\exp\left\{\lambda t\left(\phi_{\xi}(a)-1\right)\right\}$.

**Proof** 

$$\begin{aligned}\phi_{Z(t)}(a)&= \mathbb E \exp\left\{ia\sum_{k=1}^{N(t)}  \xi_k\right\}=\mathbb E_{N(t)} \mathbb E_\xi\exp\left\{ia\sum_{k=1}^{N(t)}  \xi_k\right\}
 = \mathbb E_{N(t)}\left\{\phi_{\xi}(a)^{N(t)} \right\}
 \\ &= \sum_{k=0}^\infty \frac{(\lambda t)^k}{k!}e^{-\lambda t}\phi_{\xi}(a)^{k} = \exp\left\{\lambda t\left(\phi_{\xi}(a)-1\right)\right\}.
\end{aligned}$$
