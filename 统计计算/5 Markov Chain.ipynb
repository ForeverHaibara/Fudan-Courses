## Markov Chain


(It is suggested to learn the details of Markov chain in the course \<随机过程导论 Stochastic Processes\>)

Markov chain can be formulated as below:

Let $S = \{s_1,\dotsc,s_N\}$ be a collection of states. In each timestamp $t\in\mathbb N$, a sequence $\{x_t\}$ takes its value from those states. Also, in the next timestamp $t+1$, we have the following transition rule:
$$\mathbb P(x_{t+1}=s_j|x_t=s_i) = p_{ij}.$$

Together $P = [p_{ij}]_{1\leqslant i,j\leqslant N}$ is called the transition matrix of the Markov chain. Each row of $P$ sums to $1$ and $\mathbf 1\in\mathbb R^{N\times 1}$ is thus a right eigenvector: $P\mathbf 1= \mathbf 1$.

Also, if we denote $p_0= [\mathbb P(x_0=s_1),\dotsc,\mathbb P(x_0=s_N)]\in\mathbb R^{1\times N}$ to be a <font color=red>row vector</font>. Then we have

$$p_0 P^t = [\mathbb P(x_t=s_1),\dotsc,\mathbb P(x_t=s_N)]\quad\quad (t\in\mathbb N).$$

### Stationarity

In particular, we will study ergodic (遍历的) Markov chains with stationary distribution (平稳分布) $\tilde p\in\mathbb R^{1\times N}$. Mathematically, it is the property that $\lim_{t\to\infty}P^t= \mathbf 1\tilde p$. 

**Theorem** For whatever initial distribution $p_0\in\mathbb R^{1\times N}$ that $p_0\mathbf 1 = 1$, we would have 

$$ \lim_{t\to\infty}P^t= \mathbf 1\tilde p \quad\Rightarrow\quad \lim_{t\to\infty} p_0 P^t =p_0\mathbf 1 \tilde p = \tilde p.$$

This implies $[\mathbb P(x_t=s_1),\dotsc,\mathbb P(x_t=s_N)]\rightarrow \tilde p$ for whatever $p_0$.

Also, in this case it can be shown that $\tilde p P = \tilde p$, which is a left eigenvector of $P$.

### Markov Chain Law of Large Numbers

For ergodic Markov chain with transition matrix $P$ and stationary distribution $\tilde p$, let $\{X_t\}$ be the value. Then, for function $f:\ \{s_1,\dotsc,s_N\}\to \mathbb R$, we have [1, p. 184]

$$\frac{1}{T}\sum_{t=1}^T f(X_t) \rightarrow \mathbb E_{\tilde p}f \quad\text{a.s.}$$

It can be generalized to infinite-state case when $\mathbb E_{\tilde p}|f|<\infty$.

### Markov Chain Central Limit Theorem

For ergodic Markov chain with transition matrix $P$ and stationary distribution $\tilde p$, let $\{X_t\}$ be the value. Then, for function $f:\ \{s_1,\dotsc,s_N\}\to \mathbb R$, there exists constant $\tau$ such that

$$\frac{\sqrt n}{\tau}\left(\frac{1}{T}\sum_{t=1}^T f(X_t) -  \mathbb E_{\tilde p}f\right)\to_d N(0,1).$$

It can be generalized to infinite-state case when $\mathbb E_{\tilde p}|f|^2<\infty$.

```python
import numpy as np
transition = np.array([[0.1, 0.7, 0.2], [0.2, 0.5, 0.3], [0.6, 0.1, 0.3]])

def stationary_distribution(transition):
    """Return the stationary distribution of a transition matrix."""
    eigs = np.linalg.eig(transition.T)
    eigvec = list(filter(lambda x: abs(x[0]-1)<1e-5, zip(eigs[0], eigs[1].T)))[0][1]
    return np.real(eigvec / eigvec.sum()).reshape(1,-1)

eigvec = stationary_distribution(transition)
f = np.array([1., 2.5, 1.2])

# simulate a Markov chain
def next_state(x, transition):
    """Return the next state of a Markov chain."""
    return np.random.choice(transition.shape[0], p=transition[x])

np.random.seed(0)
x = [0]
for time in range(2000):
    x.append(next_state(x[-1], transition))
x = np.array(x)
avg = f[x].cumsum() / np.arange(1, x.size + 1)
expectation = np.dot(eigvec, f)


from matplotlib import pyplot as plt
plt.figure(figsize=(9, 5))
plt.plot(avg, 'b-', label = 'sample average')
plt.plot([expectation] * x.size, 'r--', label = 'stationary expectation')
[plt.xlabel('time'), plt.ylabel('average'), plt.legend(loc='lower right')]
plt.show()
```

## Metropolis–Hastings Algorithm

Recall the acceptance-rejection method for generating random variables. If we have a target distribution $f$, we can find arbitrary distribution $g$ such that $f(x)\leqslant cg(x)$ for some constant $c$. Then we can generate $X\sim g$ and accept it with probability $\frac{f(X)}{cg(X)}$. The accepted $X$ will be distributed as $f$.

However, the constant $c$ might be hard to determine. Even though we can use very large $c$, it will lead to low acceptance rate, which is not efficient.

<br>

Metropolis-Hastings algorithm is a similar method for generating random variables with target distribution $f$. It constructs a Markov chain so that the stationary distribution is $f$. The method is as follows.

First we find arbitrary conditional Markov transition kernel $Q(x'|x)$. Then we can generate $x_{t+1}$ from $Q(x'|x_t)$ and accept it with probability $\alpha$, where

$$\alpha= \min\left\{1,\frac{f(x_{t+1})Q(x_t|x_{t+1})}{f(x_t)Q(x_{t+1}|x_t)}\right\}.$$

Then we have a Markov chain $x_1,x_2,\dotsc$, and it converges in distribution to $f$.

In particular, when $Q(x'|x) = Q(x|x')$ is symmetric,

$$\alpha= \min\left\{1,\frac{f(x_{t+1})}{f(x_t)}\right\}.$$

```python
import numpy as np

def metropolis_hastings(f, Q, genQ, x0 = 0, N = 1000, seed = 0, include_reject = False):
    """
    Return a sample path from the distribution with density f using the
    Metropolis-Hastings algorithm. Q(x,y) is a transition kernel Q(x|y),
    genQ is a function that generates random variables from Q, x0 is the initial
    state, and N is the number of samples to generate.

    If include_reject is True, then the rejected proposals are included in the
    output.
    """
    np.random.seed(seed)
    x = [x0]
    while len(x) < N:
        x1, x2 = x[-1], genQ(x[-1])
        a = f(x2) * Q(x1, x2) / (f(x1) * Q(x2, x1))
        a = min(a, 1)
        accept = np.random.uniform() < a
        if accept:
            x.append(x2)
        elif include_reject:
            x.append(x1)
    return np.array(x)

def clear_reject(trace):
    """Remove the rejected proposals from a trace."""
    cond = np.concatenate([np.array([True]), trace[1:] != trace[:-1]])
    return np.compress(cond, trace)
```

```python
import numpy as np
from scipy.stats import gamma

# generate Rayleigh random variables
rayleigh = lambda var: (lambda x: x/var*np.exp(-x**2/(2*var)))

# Metropolis-Hastings algorithm using Q(x|y) = Gamma(y/2, 1/2)
Q = lambda x, y: gamma.pdf(x, y/2, scale=2)
genQ = lambda y: gamma.rvs(y/2, scale=2)


def analyze_trace(trace, f = None):
    from matplotlib import pyplot as plt
    cleared = clear_reject(trace)
    print('Rejection rate: %f' % (1 - cleared.size / trace.size))
    plt.figure(figsize=(10, 8))
    plt.subplot(211)
    plt.plot(trace)
    plt.title('Trace')
    plt.subplot(212)
    plt.hist(cleared, bins=50, density=True)
    if f is not None:
        x = np.linspace(cleared.min(), cleared.max(), 100)
        plt.plot(x, f(x), 'r--')
    plt.title('Histogram')
    plt.show()

trace = metropolis_hastings(rayleigh(2), Q, genQ, x0 = 1, N=1000, include_reject=True)
analyze_trace(trace, f = rayleigh(2))
```

### Random Walk Metropolis-Hastings

As an special case, if we require the proposal distribution $g(\cdot|x_t)$ to be an isotropic distribution: $g(x|x_t) = g(|x - x_t|)$, then it is equivalent to the following:

$$x_{t+1} = x_t + \Delta x_t \quad\quad \Delta x_t\sim g(\cdot | 0).$$

This is called random walk Metropolis-Hastings algorithm.

A common choice of the isotropic distribution is Gaussian distribution $g(x|x_t) = N(x_t,\sigma^2)$. It is suggested that the variance $\sigma^2$ be tuned so that the acceptance rate is around $[0.15,0.5]$ to achieve the best efficiency.

```python
import numpy as np

def random_walk_metropolis_hastings(f, var, **kwargs):
    """
    Return a sample path from the distribution with density f using the
    random walk Metropolis-Hastings algorithm. Parameter `var` is the variance
    of the proposed Gaussian. Fore more details please refer to the function
    `metropolis_hastings`.
    """
    # note that the Q is actually unused because Q is symmetric and Q(x'|x) = Q(x|x')
    Q = lambda x, y: np.exp(-0.5 * (y - x)**2 / var) / np.sqrt(2 * np.pi * var)
    genQ = lambda x: x + np.random.normal(scale=np.sqrt(var))
    return metropolis_hastings(f, Q, genQ, **kwargs)


from scipy.stats import t as student_t
from matplotlib import pyplot as plt

# generate t-distribution with random walk metropolis-hastings
def t_distribution(dof):
    return lambda x: student_t.pdf(x, dof)

# try different variance of the proposal distribution
variances = [.05, .5, 2, 16]
plt.figure(figsize=(10, 8))
dof = 4
records = {}
for i, var in enumerate(variances):
    trace = random_walk_metropolis_hastings(t_distribution(dof), var, x0=25, N=2000, include_reject=True)
    trace_cleared = clear_reject(trace)
    records[var] = {'trace': trace, 'rejection rate': 1 - trace_cleared.size / trace.size}
    print('var = %s rejection rate = %f' % (str(var).ljust(5,' '), records[var]['rejection rate']))
    plt.hist(trace_cleared, bins=50, density=True, alpha=0.5, label='var = %s' % var)

xx = np.linspace(-5, 30, 1000)
plt.plot(xx, t_distribution(4)(xx), 'r--', label='true density')
[plt.legend(), plt.title('Random Walk Metropolis-Hastings Histogram')]
plt.show()


plt.figure(figsize=(10, 10))
for i, var in enumerate(variances):
    plt.subplot(2,2,i+1)
    plt.plot(records[var]['trace'], alpha=0.5, color=plt.rcParams['axes.prop_cycle'].by_key()['color'][i])
    # 95% confidence interval
    [plt.plot([student_t.ppf(ci, dof)] * len(records[var]['trace']), 'k--') for ci in [.025, .975]]
    plt.title('var = %s' % var)
plt.show()
```

### Independence Sampler

Independence sampler is another special case of Metropolis-Hastings algorithm. We require $g(x|x_t) = g(x)$ be independent of $x_t$. However, it is oftentimes useless.


## Gibbs Sampler

Gibbs sampler generates high-dimensional random variables. So we need to generate $X=[X_1,\dotsc,X_d]\in\mathbb R^d$ which follows a multivariate distribution $f(x_1,\dotsc,x_d)$. In each loop, we update $x^{(t)}$ from previous value $x^{(t-1)}$ by:

$$\left\{\begin{array}{l}x_1^{(t)} \sim f(x_1|x_2^{(t-1)},\dotsc,x_d^{(t-1)})
\\ x_2^{(t)} \sim f(x_2|x_1^{(t)},x_3^{(t-1)},\dotsc,x_d^{(t-1)})
\\ \vdots \\ x_d^{(t)} \sim f(x_d|x_1^{(t)},\dotsc,x_{d-1}^{(t)})\end{array}\right.$$

Then we take $x^{(t)} = [x_1^{(t)},\dotsc,x_d^{(t)}]$.

```python
import numpy as np

def gibbs_sampler(f, x0 = 0, n = 1000, seed = 0):
    """
    Return a sample path by Gibbs sampler. 
    Parameter `x0` is the initial state and `n` is the number
    of samples to generate. Function `f` should be a rv generator taking in
    two arguments, the first is the index of the variable to be conditioned on
    and the second is the current state, with the same shape as `x0`.
    """
    if seed is not None:
        np.random.seed(seed)
    x = [x0]
    for i in range(n):
        new_x = x[-1].copy()
        for d in range(new_x.size):
            new_x[d] = f(d, new_x)
        x.append(new_x)
    return np.array(x)

def conditional_gaussian(mu, cov, d, x):
    """
    Return the conditional mean and variance for the conditional distribution of
    a multivariate Gaussian distribution. `mu` and `cov` are the mean and
    covariance matrix of the multivariate Gaussian distribution. `d` is the
    index (int) of the variable to be conditioned on and `x` is the current state.
    """
    cov_xx = cov[d, d]
    delete_d = np.concatenate([np.arange(d), np.arange(d+1, x.size)])
    cov_xy = cov[d, delete_d]
    cov_yy = cov[delete_d, :][:, delete_d]
    mu_x = mu[d]
    mu_y = mu[delete_d]
    y = x[delete_d]
    mean2 = mu_x + cov_xy @ np.linalg.solve(cov_yy, y - mu_y)
    var2 = cov_xx - cov_xy @ np.linalg.solve(cov_yy, cov_xy.T)
    return mean2, var2

def cond_gaussian_for_gibbs(mu, cov):
    def rv(d, x):
        mean2, var2 = conditional_gaussian(mu, cov, d, x)
        return np.random.normal(mean2, np.sqrt(var2))
    return rv
```

```python
import numpy as np

mu = np.array([0., 2.])
cov = np.array([1., -.75*.5, -.75*.5, .5**2]).reshape(2,2)

data = gibbs_sampler(cond_gaussian_for_gibbs(mu, cov), x0 = mu, n = 5000)
data = data[1000:] # burn-in, discard the leading 1000 samples


def analyze_samples(data, plot_ratio = .25):
    stats = {
        'Mean': data.mean(axis=0),
        'Cov': np.cov(data.T, ddof=1),
        'Corr': np.corrcoef(data.T),
    }
    for k, v in stats.items():
        [print(k + ':'), print(v), print()]

    if data.shape[1] == 2:
        from matplotlib import pyplot as plt
        from matplotlib.patches import Ellipse
        plt.figure(figsize=(10, 8))
        # note that the sample path is correlated chronologically
        # we select part of samples randomly
        np.random.seed(0)
        plot_num = round(data.shape[0] * plot_ratio)
        plot_inds = np.random.choice(data.shape[0], plot_num, replace=False)
        plot_data = data[plot_inds]
        plt.scatter(plot_data[:,0], plot_data[:,1], alpha=0.5)

        # plot the mean
        plt.scatter(mu[0], mu[1], marker='x', color='r', s=100, label='true mean')

        # plot the 95% confidence interval ellipse
        eigvals, eigvecs = np.linalg.eigh(stats['Cov'])
        angle = np.degrees(np.arctan2(*eigvecs[:,0][::-1]))
        z = 1.96
        ellipse = Ellipse(stats['Mean'], width=2*z*eigvals[0]**.5, height=2*z*eigvals[1]**.5, 
                        angle=angle, edgecolor='darkviolet', fc='None', lw=2, linestyle='--')
        plt.gca().add_patch(ellipse)
        plt.show()

analyze_samples(data)
```
