# Honor 3 Multivariate Normal

Given $X = [X_1,\dotsc,X_p]^T\in\mathbb R^p$ and if $X\sim N(\mu,\Sigma)$ we mean it has joint density
$$f(X) = \frac{1}{\sqrt{2\pi}^p |\Sigma|^\frac 12}\exp\{-\frac 12(x - \mu)^T\Sigma^{-1}(x -\mu)\}.$$
And it is well known that 
$$\Sigma = \left[\begin{matrix}\sigma_{11} & \sigma_{12} & \dotsc & \sigma_{1n}\\ 
\sigma_{21} & \sigma_{22} & \dotsc & \sigma_{2n}\\ 
\vdots & \vdots & \ddots & \vdots \\ 
\sigma_{n1} & \sigma_{n2} & \dotsc & \sigma_{nn}\end{matrix}\right]$$
where $\sigma_{ij} = {\rm Cov}(X_i,X_j)\leqslant \sqrt{\sigma_{ii}\sigma_{jj}}$ and 
$\Sigma$ is positive semidefinite.

### Affine Transformation

$$X\sim N(\mu,\Sigma)\quad \Rightarrow \quad AX+b\sim N(A\mu+b,A\Sigma A^T)$$

A direct corollary is that, $X_j\sim N(\mu_j,\sigma_{jj})$  as an marginal distribution.

### Conditional Distribution

Suppose $X = [X_1,\dotsc,X_p]^T$, $Y = [Y_1,\dotsc,Y_q]^T$ are two multivariate normal distribution. To study the relation between $X,Y$, we merge them by $Z\sim N(\mu,\Sigma)$ where the parameters can be partitioned by

$$\mu = \left[\begin{matrix}\mu_x \\ \mu_y\end{matrix}\right]\quad\quad 
\Sigma = \left[\begin{matrix}\Sigma_{XX} & \Sigma_{XY} \\ \Sigma_{YX} & \Sigma_{YY}\end{matrix}\right].$$

Then 
$$X|y  \sim N(\mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y - \mu_Y),  \ \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX} ).$$

<br>

Proof: 
$$f_{X|Y}(x|y) = \frac{f_{XY}(x,y)}{f_Y(y)} = 
\frac{\frac{1}{\sqrt{2\pi }^{p+q}|\Sigma|^\frac12}\exp\{-\frac 12([x^T,y^T]-\mu^T)\Sigma^{-1}([x^T,y^T]^T-\mu)\}}{
\frac{1}{\sqrt{2\pi }^{q}|\Sigma_{YY}|^\frac12}\exp\{-\frac 12(y^T-\mu_Y^T)\Sigma_{YY}^{-1}(y-\mu_Y)\}
}$$

Now we note that 
$$\Sigma^{-1} = 
\left(\left[\begin{matrix}I&\Sigma_{XY}\Sigma_{YY}^{-1} \\ 
  & I\end{matrix}\right]\left[\begin{matrix}\Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX} & \\ 
  & \Sigma_{YY}\end{matrix}\right]\left[\begin{matrix}I& \\ 
  \Sigma_{YY}^{-1}\Sigma_{YX}& I\end{matrix}\right]\right)^{-1}$$
So $|\Sigma|^\frac12 =|\Sigma_{YY}|^\frac12 | \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}|^\frac12$ and 
$$\begin{aligned}&\quad -\frac 12([x^T,y^T]-\mu^T)\Sigma^{-1}([x^T,y^T]^T-\mu)\\ & 
=-\frac 12([x^T,y^T]-\mu^T)\left[\begin{matrix}I& \\ 
  -\Sigma_{YY}^{-1}\Sigma_{YX}& I\end{matrix}\right]\left[\begin{matrix}\left(\Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX} \right)^{-1}& \\ 
  & \Sigma_{YY}^{-1}\end{matrix}\right]\left[\begin{matrix}I&-\Sigma_{XY}\Sigma_{YY}^{-1} \\ 
  & I\end{matrix}\right]([x^T,y^T]^T-\mu)\\   & 
=-\frac 12(x-\mu_X -\Sigma_{XY} \Sigma_{YY}^{-1} ( y - \mu_Y))^T\left(\Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX} \right)^{-1}(x-\mu_X -\Sigma_{XY} \Sigma_{YY}^{-1}( y - \mu_Y))\\ &\quad\quad  -\frac 12(y^T-\mu_Y^T)\Sigma_{YY}^{-1}(y-\mu_Y).\end{aligned}$$


#### Independence

For multivariate normal distribution
$${\rm Cov}(X,Y) = \Sigma_{XY} = 0\quad\Leftrightarrow\quad 
X|y \sim N(\mu_X,\Sigma_{XX})\quad \Leftrightarrow\quad X,Y{\rm\ independent}$$

### Maximum Likelihood Estimation

Given observations $\{X_i = [X_{i1},\dotsc,X_{ip}]^T\in\mathbb R^p\}$ where $X_i\stackrel{iid.}{\sim }N(\mu,\Sigma)$, we get the log likelihood by 
$$I(\mu,\theta) = -\frac{np}{2}\log (2\pi) - \frac n 2\log |\Sigma| -\frac 12\sum_{j=1}^n (X_i -\mu)^T\Sigma^{-1} (X_i - \mu).$$

Take the partial derivative, 
$$0 = \frac{\partial I}{\partial \mu} = -\sum_{j=1}^n (X_i - \mu)^T\Sigma^{-1}\quad \Rightarrow\quad \hat \mu = \overline X
$$
and 
$$0 = \frac{\partial I}{\partial \Sigma^{-1}} = \frac n2\Sigma-
\frac 12\sum_{j=1}^n (X_i -\hat \mu)^T (X_i -\hat \mu)\quad \Rightarrow\quad \hat \Sigma = \frac 1n\sum_{j=1}^n (X_i -\overline X)^T (X_i - \overline X).
$$

### Related Distributions

#### Wishart's Distribution

Suppose we have independent random variables $\{X_i = [X_{i1},\dotsc,X_{ip}]^T\in \mathbb R^p\}$ where $X_i\sim N(0,\Sigma_p)$. And we denote that
$M= \mathbf X\mathbf X^T = \sum_{i=1}^n X_iX_i^T\in \mathbb R^p$, we say that $M$ has Wishart distribution, denoted as 
$$M\sim W_p(n,\Sigma).$$

It is clearly that it degenerates to a chi-squared distribution when $p = 1$. Actually, for fixed $a\neq 0\in\mathbb R^p$, we have that 
$$\frac{a^TMa}{a^T\Sigma a}\sim \chi_n^2$$
and when $n\geqslant p$, 
$$\frac{a^T\Sigma^{-1}a }{a^TM^{-1}a}\sim \chi_{n-p+1}^2.$$


Remark: <font color=red>Note that we require $X_i$ has **zero** mean!</font>

#### Hotelling's T-squared Distribution

Using the notation above, if $Y\in \mathbb R^p\sim N_p(\mu, \Sigma)$ is another random variable and is independent with $M$, then we write
$$n ( y - \mu)^TM^{-1}(y - \mu)\sim T^2(p,n)$$

It degenerates to the square of a T distribution when $p = 1$.
It connects the F distribution by 
$$T^2(p,n) =\frac{np}{n-p+1}F_{p,n-p+1}.$$


#### Wilk's Lambda Distribution

Suppose $A\sim W_p(m,\Sigma)$ and $B\sim W_p(n,\Sigma)$. When $m\geqslant p$, then we write
$$\lambda = \frac{{\rm det}(A)}{{\rm det}(A+B)} \sim \Lambda (p,m,n).$$

When one of $p,m,n$ is $1$, it is related to F distribution. For example 
$$\frac{1 - \Lambda(p,m,1)}{\Lambda (p,m,1)}\sim \frac{p}{m - p+1}F_{p,m-p+1}.$$

### Hypothesis Testing

#### T Test

Similar as the T test $t = \frac{\sqrt n(\overline X - \mu)}{\widehat {\rm SE}(X)}$ or
 $t^2 = n(\overline X - \mu){\widehat {\rm SE}(X)}^{-1}(\overline X - \mu)$ of the univariate normal distribution, we test Hotelling's $T$ or $T^2$

$$T^2 =n (\overline X - \mu)\hat \Sigma^{-1}(\overline X - \mu)$$
where $\hat\Sigma = \frac{1}{n-1}\sum_{j=1}^n (X_i- \overline X)(X_i - \overline X)^T$ is the unbiased sample variance.

Recall that $T^2 \sim T^2(p,n-1)\sim \frac{(n-1)p}{n-p}F_{p,n-p}$ where $p$ is the dimension of each datum, we reject $H_0$ by refering to $F$,
$$T^2 > \frac{(n-1)p}{n-p}F_{p,n-p}(\alpha)\quad \Rightarrow\quad {\rm reject\ }H_0.$$
