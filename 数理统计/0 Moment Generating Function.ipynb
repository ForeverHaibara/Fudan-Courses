# 0 Moment Generating Function (MGF)

For single random variable $X$ (not a vector), the moment generating function $\psi_X(t)$ is a function with regard to $t$. 

$$\psi_X(t) = \mathbb E(e^{tX})$$

Properties.

1. The k-th derivative at zero is the k-th moment. $\psi_X^{(k)}(0) = \mathbb E(x^k) = \mu_k$.
2. If $Y = a + bX$, $\psi_Y(t) \equiv e^{at}\psi_X(bt)$. 
3. If $X_1,\dotsc,X_n$ are independent, $\psi_{\sum X_i}(t) = \prod \psi_{X_i}(t)$. 

## Characteristic Function (CF)

One can also take complex values $t$ in the MGF, which is in particular called a characteristic function.
$$\hat \mu_X(t) = \psi_X(it) = \mathbb E(e^{itX})$$

## Normal Distribution

### Moment Generating Function

The moment generating function of a standard normal distribution $X\sim N(0,1)$ is 
$$\psi_X(t) = \mathbb E(e^{tX}) = \int_{-\infty}^{\infty} \frac{1}{\sqrt {2\pi}}e^{tx-\frac{x^2}{2}}dx
=e^{\frac {t^2}{2}}\int_{-\infty}^{\infty} \frac{1}{\sqrt {2\pi}}e^{-\frac{(x-t)^2}{2}}dx=e^{\frac {t^2}{2}}\ .
$$

In general if we have $X\sim N(\mu,\sigma^2)$ we have
$$
\psi_X(t) = \int_{-\infty}^{\infty} \frac{1}{\sqrt {2\pi}|\sigma|}e^{t x-\frac{(x-\mu)^2}{2\sigma^2}}dx
=e^{\mu t + \frac{\sigma ^2t^2}{2}}\ .
$$

### Characteristic Function

The characteristic function of a standard normal distribution $X\sim N(0,1)$ is 
$$\psi_X(it) = \mathbb E(e^{itX}) = \int_{-\infty}^{\infty} \frac{1}{\sqrt {2\pi}}e^{itx-\frac{x^2}{2}}dx
=e^{-\frac {t^2}{2}}\ .
$$

Proof: We integral the function $e^{-\frac{z^2}{2}}$ along a rectangle contour with sides 

$$\gamma_1: [-R,R],\ \gamma_2: [R,R-it],\  
\gamma_3: [R- it,-R- it],\ \gamma_4: [-R- it,-R].$$ 

Clearly that $\int_{\gamma_2}$ and $\int_{\gamma_4}$ vanish
as $R\rightarrow +\infty$ so
$$\int_{-\infty}^{\infty} \frac{1}{\sqrt {2\pi}}e^{itx-\frac{x^2}{2}}dx
=e^{-\frac{t^2}{2}}\int_{-\infty}^{\infty} \frac{1}{\sqrt {2\pi}}e^{-\frac{(x-it)^2}{2}}dx
=e^{-\frac{t^2}{2}}\left( -\int_{\gamma_3}\right)=e^{-\frac{t^2}{2}}\int_{\gamma_1}=e^{-\frac{t^2}{2}}
$$

<br>

In general if we have $X\sim N(\mu,\sigma^2)$ we have
$$
\psi_X(it) = \int_{-\infty}^{\infty} \frac{1}{\sqrt {2\pi}|\sigma|}e^{it x-\frac{(x-\mu)^2}{2\sigma^2}}dx
=e^{i\mu t - \frac{\sigma^2t^2}{2}}\ .
$$

## Inversion


### Inversion
If $\mathbb P(X\leqslant t)$ is continuous at $t = a$ and $t=b$ and $a<b$, one can restore the distribution by MGF, 
$$\mathbb P(a\leqslant X\leqslant b) = p.v. \ \frac{1}{2\pi i}\int_{-\infty}^{\infty } \frac{e^{-iat} - e^{-ibt}}{t}\psi(it)dt.$$

Proof: 
$$\begin{aligned}
p.v. \ \frac{1}{2\pi i}\int_{-\infty}^{\infty } \frac{e^{-iat} - e^{-ibt}}{t}\psi(it)dt
& = p.v. \ \frac{1}{2\pi i}\int_{-\infty}^{\infty } \frac{e^{-iat} - e^{-ibt}}{t}\mathbb E(e^{itX})dt\\ 
& = \mathbb E \left( p.v. \ \frac{1}{2\pi i}\int_{-\infty}^{\infty } \frac{e^{it(X-a)} - e^{it(X-b)}}{t}dt\right)\\
& = \mathbb E \left( \ \frac{1}{2\pi i}\int_{0}^{\infty } \frac{e^{it(X-a)}-e^{-it(X-a)}- e^{it(X-b)}+e^{-it(X-b)}}{t}dt\right)\\
& = \mathbb E \left( \ \frac{1}{\pi }\int_{0}^{\infty } \frac{\sin(t(X-a)) - \sin(t(X-b))}{t}dt\right)
\end{aligned}
$$
The well-known Dirichlet integral claims that when $\theta \neq 0$,
$$\int_0^{\infty}\frac{\sin (\theta t)}{t}dt = \int_0^{\infty}\frac{\sin (\theta t)}{|\theta| t}|\theta| dt =
\mathbb {\rm sgn}(\theta)\int_0^{\infty} \frac{\sin u}{u}du =  
\mathbb {\rm sgn}(\theta)\frac{\pi}{2}.$$
Remove the singularity at $x = a$ and $x = b$, we attain
$$\begin{aligned}
p.v. \ \frac{1}{2\pi i}\int_{-\infty}^{\infty } \frac{e^{-iat} - e^{-ibt}}{t}\psi(it)dt
& = \mathbb E \left( \ \frac{1}{\pi }\int_{0}^{\infty } \frac{\sin(t(X-a)) - \sin(t(X-b))}{t}dt\right)\\
& =\mathbb E \left( \ \mathbb I_{a\leqslant X\leqslant b}\right) = \mathbb P(a\leqslant X\leqslant b) \ .
\end{aligned}
$$

### Expectation

For a generlization of the inversion, one can calculate the expectation of 
$\varphi(x)$, a continuous function and bounded over $\mathbb R$, by,
$$ \mathbb E(\varphi(X)) = \lim_{u\rightarrow 0_+}\frac{1}{2\pi}\int_{-\infty}^{\infty} \varphi(t)
\int_{-\infty}^{\infty} \psi(iz)e^{-itz-\frac{uz^2}{2}}dzdt.$$

Proof: 
$$\begin{aligned}\mathbb E(\varphi(X)) & = \lim_{u\rightarrow 0_+}\frac{1}{2\pi}\int_{-\infty}^{\infty} \varphi(t)
\int_{-\infty}^{\infty} \mathbb E(e^{izX})e^{-itz-\frac{uz^2}{2}}dzdt\\
& =  \lim_{u\rightarrow 0_+}\mathbb E\left(\frac{1}{2\pi}\int_{-\infty}^{\infty} \varphi(t) \int_{-\infty}^{\infty} e^{i(X-t)z-\frac{uz^2}{2}}dzdt\right)
\end{aligned}$$
To compute the integral $\int_{-\infty}^{\infty} e^{i(X-t)z-\frac{uz^2}{2}}dz$, we construct $Y\sim N(0,\frac{1}{ u})$ and use the result above to conclude
$$\frac{1}{\sqrt{2\pi }\frac{1}{\sqrt u}}\int_{-\infty}^{\infty} e^{i(X-t)z-\frac{uz^2}{2}}dz
=\psi_Y(i(X-t))=e^{-\frac{(X-t)^2}{2u}}\ .
$$
Therefore, 
$$\begin{aligned}
\lim_{u\rightarrow 0_+}\frac{1}{2\pi}\int_{-\infty}^{\infty} \varphi(t)
\int_{-\infty}^{\infty} \psi(iz)e^{-itz-\frac{uz^2}{2}}dzdt 
=& \lim_{u\rightarrow 0_+}\mathbb E\left(\int_{-\infty}^{\infty} \varphi(t)\cdot \frac{1}{\sqrt{2\pi u}} e^{-\frac{(X-t)^2}{2u}}dt\right)\\
\xlongequal{t = X+\sqrt u r} & \lim_{u\rightarrow 0_+}\mathbb E\left(\int_{-\infty}^{\infty} \varphi(X+\sqrt ur)\cdot \frac{1}{\sqrt{2\pi }} e^{-\frac{r^2}{2}}dr\right)\\
\xlongequal{Lebesgue} &\  \mathbb E\left(\int_{-\infty}^{\infty} \lim_{u\rightarrow 0_+} \varphi(X+\sqrt ur)\cdot \frac{1}{\sqrt{2\pi }} e^{-\frac{r^2}{2}}dr\right)
\\ = &\  \mathbb E(\varphi (X))\ .
\end{aligned}$$
This completes the proof.
